<details>
<summary><b>Ver solucion - Ejercicio 4</b></summary>

Cada metodo devuelve un **nuevo string** (los strings en Python son inmutables, no se modifican en su sitio).

```python
# --- Solucion Ejercicio 4: Strings y sus metodos ---

frase = "A quien madruga, dios le ayuda"

# CONCEPTO: Los strings en Python son INMUTABLES: los metodos NO modifican el original,
# sino que devuelven un NUEVO string. La variable 'frase' nunca cambia.

print(frase.upper())   # .upper() -> todo a MAYUSCULAS: "A QUIEN MADRUGA, DIOS LE AYUDA"
print(frase.lower())   # .lower() -> todo a minusculas: "a quien madruga, dios le ayuda"
print(frase.title())   # .title() -> primera letra de cada palabra en mayuscula

# CONCEPTO: .split() divide un string en una LISTA usando un separador
# Sin argumento, separa por espacios (y agrupa espacios multiples)
palabras = frase.split()  # Resultado: ['A', 'quien', 'madruga,', 'dios', 'le', 'ayuda']
print(palabras)

# CONCEPTO: .replace(viejo, nuevo) sustituye TODAS las apariciones del texto viejo por el nuevo
print(frase.replace(",", ";"))  # Cambia comas por puntos y comas

# .replace("a", "") reemplaza "a" por nada, es decir, ELIMINA todas las "a" minusculas
# La "A" mayuscula NO se ve afectada porque Python distingue mayusculas de minusculas
print(frase.replace("a", ""))
```

</details>

<details>
<summary><b>Ver solucion - Ejercicio 6</b></summary>

El slicing usa la sintaxis `variable[inicio:fin:paso]`. Los indices negativos cuentan desde el final.

```python
# --- Solucion Ejercicio 6: Variable con informacion personal (Slicing) ---

info = "Marina de Empresas 2025"

# CONCEPTO: len() devuelve el numero total de caracteres (incluidos espacios)
print(f"Longitud: {len(info)}")  # 23 caracteres

# CONCEPTO: Indexing - los strings se indexan desde 0 (el primer caracter es [0])
print(f"Primera letra: {info[0]}")  # 'M' (posicion 0)

# CONCEPTO: Indices negativos cuentan desde el final: -1 es el ultimo, -2 el penultimo...
# info[-4:] significa "desde 4 posiciones antes del final hasta el final"
print(f"Ultimas 4: {info[-4:]}")  # '2025'

# CONCEPTO: Slicing [inicio:fin:paso] - con paso -1 recorre al reves
# [::-1] sin inicio ni fin significa "todo el string, pero al reves"
print(f"Al reves: {info[::-1]}")  # '5202 saserpmE ed aniraM'

# CONCEPTO: .count() cuenta cuantas veces aparece un substring dentro del string
# Atencion: distingue mayusculas de minusculas ('a' != 'A')
# En "Marina de Empresas 2025": las 'a' minusculas estan en "rin_a", "mpres_a_s" = 3 veces
print(f"Veces que aparece 'a': {info.count('a')}")  # 3 (no cuenta la 'M' porque es mayuscula)
```

</details>

> **Recuerda:** Slicing - como funciona
>
> Sintaxis: **`variable[inicio:fin:paso]`**
>
> | Expresion | Resultado | Descripcion |
> |---|---|---|
> | `info[0]` | `'M'` | Primer caracter |
> | `info[-1]` | `'5'` | Ultimo caracter |
> | `info[0:6]` | `'Marina'` | Desde posicion 0 hasta 5 (el 6 NO se incluye) |
> | `info[:6]` | `'Marina'` | Igual que `info[0:6]` |
> | `info[6:]` | `' de Empresas 2025'` | Desde posicion 6 hasta el final |
> | `info[-4:]` | `'2025'` | Los ultimos 4 caracteres |
> | `info[::-1]` | `'5202 saserpmE ed aniraM'` | Todo al reves (paso -1) |

<details>
<summary><b>Ver solucion - Ejercicio 7</b></summary>

Las conversiones funcionan cuando el contenido tiene sentido para el tipo destino. Si no, Python lanza un `ValueError`.

```python
# --- Solucion Ejercicio 7: Conversion de tipos ---

# CONCEPTO: int() convierte un string que contiene SOLO digitos a un entero
# Funciona porque "42" es un numero valido; "hello" causaria ValueError
num_entero = int("42")
print(f"int('42') = {num_entero}, tipo: {type(num_entero)}")

# CONCEPTO: float() convierte un string numerico a numero decimal
num_decimal = float("3.14")
print(f"float('3.14') = {num_decimal}, tipo: {type(num_decimal)}")

# CONCEPTO: str() convierte CUALQUIER tipo a string (siempre funciona, nunca falla)
texto_num = str(100)
print(f"str(100) = '{texto_num}', tipo: {type(texto_num)}")

# CONCEPTO: bool() convierte a booleano siguiendo las reglas "truthy/falsy"
# Los valores "falsy" (que dan False): 0, 0.0, "", [], {}, None
# Todo lo demas es "truthy" (da True)
bool_cero = bool(0)        # 0 es falsy -> False
print(f"bool(0) = {bool_cero}, tipo: {type(bool_cero)}")

bool_texto = bool("hello")  # Cualquier string NO vacio es truthy -> True
print(f"bool('hello') = {bool_texto}, tipo: {type(bool_texto)}")

# --- Que pasa cuando la conversion NO tiene sentido? ---
# CONCEPTO: try/except captura errores para que el programa no se detenga
# Sin try/except, int("hello") detendria TODO el programa con un ValueError
print("\nIntentando int('hello'):")
try:
    resultado = int("hello")  # "hello" no es un numero -> ValueError
except ValueError as e:
    # 'as e' guarda el mensaje del error en la variable 'e' para poder imprimirlo
    print(f"  Error capturado: {e}")
```

</details>

> **Recuerda:** Reglas de conversion de tipos
>
> | Conversion | Resultado | Nota |
> |---|---|---|
> | `int("42")` | `42` | String numerico a entero |
> | `float("3.14")` | `3.14` | String numerico a decimal |
> | `str(100)` | `"100"` | Cualquier cosa a string (siempre funciona) |
> | `bool(0)` | `False` | `0`, `0.0`, `""`, `[]`, `{}`, `None` son `False` |
> | `bool("hello")` | `True` | Todo lo demas es `True` |
> | `int("hello")` | **ERROR** | No se puede convertir texto no numerico |

<details>
<summary><b>Ver solucion - Ejercicio 8</b></summary>

Las 4 formas producen el mismo resultado, pero los **f-strings** son el estandar moderno recomendado.

```python
# --- Solucion Ejercicio 8: Batalla de Formateo de Strings ---

nombre = "Francisco"
edad = 25
nota = 9.567

# --- Metodo 1: Concatenacion con + ---
# Necesitas str(edad) porque no se puede concatenar str + int directamente (TypeError)
# Es el metodo mas antiguo y dificil de leer con muchas variables
print("1) " + "Me llamo " + nombre + " y tengo " + str(edad) + " anios")

# --- Metodo 2: Operador % (estilo C) ---
# %s = string, %d = entero (digit), %f = float
# Los valores van en una tupla al final despues de %
print("2) Me llamo %s y tengo %d anios" % (nombre, edad))

# --- Metodo 3: .format() ---
# Las llaves {} se reemplazan en orden por los argumentos de .format()
# Mejor que %, pero sigue siendo verboso para muchas variables
print("3) Me llamo {} y tengo {} anios".format(nombre, edad))

# --- Metodo 4: f-string (Python 3.6+) - RECOMENDADO ---
# CONCEPTO: f-strings son el estandar moderno. Pones f antes de las comillas
# y metes las variables directamente dentro de {}. Mas legible y rapido.
print(f"4) Me llamo {nombre} y tengo {edad} anios")

# --- Bonus: Formateo de decimales con f-strings ---
# CONCEPTO: {variable:.Nf} formatea un float con N decimales
# :.1f = 1 decimal, :.2f = 2 decimales, :.0f = sin decimales (redondea)
print(f"Nota con 1 decimal: {nota:.1f}")    # 9.6
print(f"Nota con 2 decimales: {nota:.2f}")  # 9.57
print(f"Nota sin decimales: {nota:.0f}")    # 10 (redondea hacia arriba)
```

</details>

> **Recuerda:** Comparacion de metodos de formateo (de peor a mejor)
>
> | # | Metodo | Ejemplo | Veredicto |
> |---|---|---|---|
> | 1 | Concatenacion (`+`) | `"Hola " + nombre` | Dificil de leer, necesita `str()` |
> | 2 | Operador `%` | `"Hola %s" % nombre` | Estilo antiguo, poco flexible |
> | 3 | `.format()` | `"Hola {}".format(nombre)` | Bueno, pero verboso |
> | 4 | **f-string** (`f""`) | `f"Hola {nombre}"` | **MODERNO, legible, rapido. RECOMENDADO** |

Usamos `.lower()` para convertir a minusculas, un bucle `for` para recorrer cada caracter, `.isalpha()` para detectar letras, y `repr()` para ver caracteres ocultos.

```python
# --- Solucion Ejercicio 9.1: Convertir grito a voz baja ---

grito = "HOLA, COMO ESTAS? TODO BIEN POR AQUI!"

# CONCEPTO: .lower() devuelve un NUEVO string con todo en minusculas
# El string original 'grito' NO se modifica (los strings son inmutables)
voz_baja = grito.lower()

print(f"Grito original: {grito}")     # Sigue en mayusculas
print(f"Voz baja: {voz_baja}")        # Todo en minusculas
```


```python
# --- Solucion Ejercicio 9.2: Convertir mezcla a minusculas ---

mezcla = "HoLa MuNdO, EsTo Es PyThOn"

print(f"Mezcla original: {mezcla}")
# .lower() normaliza TODO a minusculas, sin importar la mezcla original
# Esto es muy util para comparar textos sin preocuparse por mayusculas
print(f"Normalizada: {mezcla.lower()}")
```


```python
# --- Solucion Ejercicio 9.3: Contar vocales, consonantes y otros ---

texto_analizar = "Programacion en Python 2025"
print(f"Analizando: '{texto_analizar}'")

# Definimos todas las vocales (mayusculas y minusculas) en un solo string
# para poder buscar con 'in' sin necesidad de convertir a minusculas
vocales = "aeiouAEIOU"

# Inicializamos 3 contadores a 0 antes del bucle
contador_vocales = 0
contador_consonantes = 0
contador_otros = 0

# CONCEPTO: Un bucle for recorre cada caracter del string, uno a uno
# En cada iteracion, 'caracter' toma el valor del siguiente caracter
for caracter in texto_analizar:
    # CONCEPTO: 'in' busca si un elemento existe dentro de una coleccion
    if caracter in vocales:
        contador_vocales += 1  # Es vocal -> incrementar contador
    elif caracter.isalpha():
        # CONCEPTO: .isalpha() devuelve True solo si el caracter es una letra
        # Si llegamos aqui, sabemos que es letra (isalpha) pero NO es vocal -> es consonante
        contador_consonantes += 1
    else:
        # No es letra: puede ser espacio, numero, signo de puntuacion, etc.
        contador_otros += 1

print(f"Vocales: {contador_vocales}")
print(f"Consonantes: {contador_consonantes}")
print(f"Otros (numeros, espacios, signos): {contador_otros}")
```


```python
# --- Solucion Ejercicio 9.4: Caracteres ocultos con repr() ---

# CONCEPTO: \t es un tabulador (tab) y \n es un salto de linea (newline)
# Son "caracteres de escape": ocupan 1 caracter pero se escriben con 2 simbolos
oculto = "Hola,\tmundo!\n"

# print() INTERPRETA los caracteres de escape: muestra el tab como espacio y el \n como salto
# end="" evita que print anada un segundo salto de linea (ya hay uno en \n)
print(f"print normal: {oculto}", end="")

# CONCEPTO: repr() muestra la representacion "cruda" del string,
# haciendo VISIBLES los caracteres ocultos como \t y \n
# Es muy util para depurar cuando sospechas que hay caracteres invisibles
print(f"print con repr: {repr(oculto)}")
```

> **Metodos de caso:**
>
> | Metodo | Descripcion |
> |---|---|
> | `.lower()` | Todo minusculas |

> | Metodo | Descripcion |
> |---|---|
> | `.isalpha()` | `True` si solo tiene letras |

> | `.isalnum()` | `True` si solo tiene letras y numeros |

## Errores Comunes (evita estos!)

| Error | Ejemplo incorrecto | Correccion |
|:---|:---|:---|
| Confundir `print()` con `return` | `print(x)` dentro de una funcion creyendo que devuelve el valor | `return x` para devolver, `print(x)` solo para mostrar en pantalla |
| Olvidar comillas en strings | `nombre = Hola` | `nombre = "Hola"` -- los strings siempre van entre comillas |
| Concatenar str + int sin convertir | `"Tengo " + 25 + " anios"` -> TypeError | `"Tengo " + str(25) + " anios"` o mejor: `f"Tengo {25} anios"` |
| Confundir `len()` (cuenta desde 1) con indice (empieza en 0) | `texto = "Hola"` -> `texto[4]` -> IndexError | `len("Hola")` es 4, pero el ultimo indice es 3: `texto[3]` o `texto[-1]` |
| Slicing: olvidar que el final no se incluye | `"Python"[0:3]` esperando `"Pyth"` | `"Python"[0:3]` devuelve `"Pyt"` (posiciones 0, 1, 2). Para incluir la 3: `[0:4]` |
