- Una **letra** (string de un caracter)

> **Tip:** convierte todo a minusculas con `.lower()`.

### Tu turno

<details>
<summary><b>Ver solucion - Ejercicio 3</b></summary>

Usamos `.lower()` para normalizar y `.count()` para contar apariciones. El metodo `.count()` cuenta cuantas veces aparece un substring en un string. Tambien funciona con listas: `[1,2,3,2].count(2)` devuelve `2`.

```python
# --- Funcion para contar una letra en un texto ---

def contar_letra(texto, letra):
    # Convertimos ambos a minusculas para que la comparacion sea case-insensitive.
    # CONCEPTO: .lower() devuelve una NUEVA cadena en minusculas (no modifica la original,
    # porque los strings en Python son inmutables).
    texto_lower = texto.lower()
    letra_lower = letra.lower()

    # CONCEPTO: str.count(sub) cuenta cuantas veces aparece el substring 'sub'
    # dentro del string. Es un metodo nativo de Python, no necesita bucle manual.
    return texto_lower.count(letra_lower)


# --- Pruebas ---
print(contar_letra("Hola Mundo", "o"))  # 2 (la 'o' aparece en "hola" y "mundo")
print(contar_letra("Python", "P"))      # 1 (tras .lower(), "python".count("p") = 1)
```

## Ejercicio 4 de 18: Funcion - Conteo de letras (diccionario) | Facil

> **Conceptos clave:** `def` | diccionario | `.isalpha()` | `.lower()` | patron contador

Crea una funcion `contar_todas_letras` que reciba un texto y devuelva un diccionario con cada letra y cuantas veces aparece.

- Convierte a minusculas
- Ignora espacios y signos (solo cuenta letras con `.isalpha()`)

**Ejemplo:** `"Hola"` -> `{'h': 1, 'o': 1, 'l': 1, 'a': 1}`

### Tu turno

````python
# TU CODIGO AQUI




# Prueba:
# print(contar_todas_letras("Hola Mundo"))
# Esperado: {'h': 1, 'o': 2, 'l': 1, 'a': 1, 'm': 1, 'u': 1, 'n': 1, 'd': 1}
````

<details>
<summary><b>Ver solucion - Ejercicio 4</b></summary>

**Patron muy comun** para contar apariciones con diccionario:

1. Crear diccionario vacio
2. Recorrer los elementos
3. Si el elemento ya esta como clave -> sumar 1
4. Si no esta -> crearlo con valor 1

```python
# --- Funcion para contar TODAS las letras con diccionario ---

# CONCEPTO: Patron clasico "contador con diccionario" - uno de los mas usados en Python.
# Recorremos cada elemento, y usamos un dict para llevar la cuenta de apariciones.
def contar_todas_letras(texto):
    texto_lower = texto.lower()  # Normalizamos a minusculas
    conteo = {}                  # Diccionario vacio donde guardaremos letra: cantidad

    # Recorremos cada caracter del texto uno por uno
    for caracter in texto_lower:
        # .isalpha() devuelve True solo si el caracter es una letra (a-z).
        # Esto filtra espacios, numeros, signos de puntuacion, etc.
        if caracter.isalpha():
            if caracter in conteo:
                # Si la letra ya esta en el diccionario, incrementamos su contador
                conteo[caracter] += 1
            else:
                # Si es la primera vez que vemos esta letra, la inicializamos en 1
                conteo[caracter] = 1

    return conteo


# --- Prueba ---
# Esperado: {'h': 1, 'o': 2, 'l': 1, 'a': 1, 'm': 1, 'u': 1, 'n': 1, 'd': 1}
print(contar_todas_letras("Hola Mundo"))
```

<details>
<summary><b>Ver solucion - Ejercicio 5</b></summary>

Cada funcion calcula un area especifica. Luego combinamos llamadas para resolver los problemas.

- `:.2f` formatea el numero a 2 decimales en un f-string
- `import math` importa el modulo con funciones matematicas: `math.pi`, `math.sqrt()`, `math.floor()`, `math.ceil()`, etc.

```python
# --- Funciones de areas geometricas ---

import math  # Importamos math para usar math.pi (3.141592...)


# Cada funcion resuelve UNA tarea especifica (principio de responsabilidad unica).
# Esto permite reutilizarlas en cualquier combinacion.
def area_cuadrado(lado):
    return lado * lado  # Area = lado^2


def area_triangulo(base, altura):
    return (base * altura) / 2  # Area = (base * altura) / 2


def area_circulo(radio):
    # CONCEPTO: ** es el operador de potencia en Python. radio ** 2 = radio al cuadrado.
    # math.pi da el valor de pi con 15 decimales de precision.
    return math.pi * radio ** 2  # Area = pi * r^2


# --- a) 2 circulos de radio 10 + 1 triangulo de base 3 y altura 7 ---
# Llamamos a las funciones y combinamos sus resultados con operadores aritmeticos.
resultado_a = 2 * area_circulo(10) + area_triangulo(3, 7)
# CONCEPTO: :.2f dentro de un f-string formatea el numero a 2 decimales.
print(f"a) 2 circulos r=10 + triangulo 3x7 = {resultado_a:.2f}")

# --- b) cuadrado lado=10 + 3 circulos (r=4,6,6) + 5 triangulos (b=2, h=4) ---
# Usamos parentesis para partir la expresion en varias lineas (legibilidad).
resultado_b = (area_cuadrado(10)
               + area_circulo(4) + 2 * area_circulo(6)
               + 5 * area_triangulo(2, 4))
print(f"b) cuadrado + 3 circulos + 5 triangulos = {resultado_b:.2f}")
```

## Conceptos: Tipos de Argumentos en Funciones

Python ofrece varios tipos de argumentos:

| Tipo | Sintaxis | Descripcion |
|------|----------|-------------|
| **Posicional** | `def f(a, b)` | Se pasan en orden |
| **Keyword-only** | `def f(*, a, b)` | Solo se pasan por nombre |
| **Por defecto** | `def f(a=10)` | Tiene valor si no se pasa |
| **`*args`** | `def f(*args)` | Tupla con N argumentos posicionales |
| **`**kwargs`** | `def f(**kwargs)` | Diccionario con argumentos por nombre |

> **Recuerda:** el orden obligatorio de parametros es: posicionales, `*args`, keyword-only, `**kwargs`.

```python
def funcion(pos1, pos2, *args, kw1, kw2, **kwargs):
#           ^posicional  ^args  ^keyword   ^kwargs
```

### Tu turno

````python
# TU CODIGO AQUI




# Pruebas:
# print(sumar_todos(1, 3, 5, 7))      # 16
# print(sumar_todos(10, 20))           # 30
# print(sumar_todos(1, 2, 3, 4, 5))   # 15
````

<details>
<summary><b>Ver solucion - Ejercicio 8</b></summary>

`*args` recoge **todos** los argumentos posicionales en una **tupla**:

```python
sumar_todos(1, 3, 5)  # -> args = (1, 3, 5)
```

Puedes recorrerla con `for`, acceder por indice, etc. La forma corta seria `return sum(args)`.

```python
# --- Funcion con *args (argumentos variables) ---

# CONCEPTO: *args recoge TODOS los argumentos posicionales en una tupla.
# Ejemplo: sumar_todos(1, 3, 5) -> args = (1, 3, 5)
# Esto permite que la funcion acepte cualquier cantidad de argumentos.
def sumar_todos(*args):
    total = 0  # Acumulador: empieza en 0 y va sumando cada numero
    for numero in args:  # Recorremos la tupla de argumentos
        total += numero  # Equivale a total = total + numero
    return total
    # NOTA: una forma mas corta seria return sum(args), pero el bucle
    # muestra como funciona internamente.


# --- Pruebas ---
print(sumar_todos(1, 3, 5, 7))     # 16  (args = (1, 3, 5, 7))
print(sumar_todos(10, 20))          # 30  (args = (10, 20))
print(sumar_todos(1, 2, 3, 4, 5))  # 15  (args = (1, 2, 3, 4, 5))
```

</details>

### Tu turno

````python
# TU CODIGO AQUI




# Prueba:
# mostrar_detalles("Detalle 1", "Detalle 2", color="rojo", talla="M")
````

<details>
<summary><b>Ver solucion - Ejercicio 10</b></summary>

> **Recuerda:** el **orden obligatorio de parametros** es: posicionales normales, `*args`, keyword-only (tras `*`), `**kwargs`.

```python
def funcion(pos1, pos2, *args, kw1, kw2, **kwargs):
```

```python
# --- Funcion combinando *args y **kwargs ---

# CONCEPTO: *args y **kwargs pueden usarse juntos en la misma funcion.
# Orden obligatorio de parametros: posicionales, *args, keyword-only, **kwargs.
# args recoge los posicionales extras en una tupla.
# kwargs recoge los argumentos por nombre extras en un diccionario.
def mostrar_detalles(*args, **kwargs):
    # --- Imprimir detalles generales (args = tupla) ---
    print("Detalles generales:")
```

Combinamos `*args` con un parametro por defecto. Notar que `total` va **despues** de `*args`.

```python
# --- Funcion con *args y parametro keyword-only por defecto ---

# CONCEPTO: Cuando *args esta presente, los parametros que vienen DESPUES
# son automaticamente keyword-only (solo se pueden pasar por nombre).
# total=50 es keyword-only con valor por defecto.
def restar_pares(*args, total=50):
```

## Conceptos: Recursion

La **recursion** es cuando una funcion se llama a si misma. Toda funcion recursiva necesita:

1. **Caso base:** la condicion que detiene la recursion
2. **Caso recursivo:** la funcion se llama a si misma con un problema **mas pequeno**

```
factorial(4):
  return 4 * factorial(3)
    return 3 * factorial(2)
      return 2 * factorial(1)
        return 1 * factorial(0)
          return 1            <-- caso base, empieza a "volver"
        return 1 * 1 = 1
      return 2 * 1 = 2
    return 3 * 2 = 6
  return 4 * 6 = 24
```

**Recursion vs Iteracion:**

- **Recursion:** mas elegante y natural para problemas recursivos (arboles, fractales, divide y venceras)
- **Iteracion:** mas eficiente en memoria (no crea muchas llamadas en el stack)
- Python tiene un limite de recursion (~1000 llamadas por defecto)

> **Recuerda (MIT 6.0001):** La recursion es fundamental en ciencias de la computacion. Muchos algoritmos clasicos (ordenacion, busqueda, recorrido de grafos) se expresan naturalmente de forma recursiva.

Crea una funcion `factorial(n)` que calcule `n!` usando **recursion**.

**Reglas:**

- Caso recursivo: `factorial(n) = n * factorial(n - 1)`

Tambien crea una funcion `factorial_iterativo(n)` que haga lo mismo pero usando un bucle `for` (sin recursion), para comparar.

````python
# TU CODIGO AQUI




# Pruebas:
# print(factorial(5))              # 120
# print(factorial(0))              # 1
# print(factorial(-3))             # Mensaje de error
# print(factorial_iterativo(5))    # 120
````

<details>
<summary><b>Ver solucion</b></summary>

Dos implementaciones del mismo algoritmo: recursiva e iterativa.

```python
# --- Factorial: version recursiva vs iterativa ---

# CONCEPTO: Recursion = una funcion que se llama a si misma.
# Toda recursion necesita: 1) caso base (cuando parar) y 2) caso recursivo (reducir el problema).
def factorial(n):
    """Calcula n! usando recursion."""
    if n < 0:
        # Validacion: el factorial no existe para numeros negativos
        return "Error: factorial no definido para numeros negativos"
    if n == 0:
        # Caso base: factorial(0) = 1 por definicion matematica.
        # Sin caso base, la recursion seria infinita y Python daria RecursionError.
        return 1
    # Caso recursivo: n! = n * (n-1)!
    # Ejemplo: factorial(4) = 4 * factorial(3) = 4 * 3 * factorial(2) = ... = 4 * 3 * 2 * 1 = 24
    return n * factorial(n - 1)


# CONCEPTO: Misma logica pero con bucle (iterativo). Mas eficiente en memoria
# porque no crea multiples llamadas en el stack de Python.
def factorial_iterativo(n):
    """Calcula n! usando un bucle (iterativo)."""
    if n < 0:
        return "Error: factorial no definido para numeros negativos"
    resultado = 1  # Acumulador: empieza en 1 porque multiplicar por 1 no cambia nada
    # range(1, n+1) genera: 1, 2, 3, ..., n
    for i in range(1, n + 1):
        resultado *= i  # Equivale a resultado = resultado * i
    return resultado


# --- Pruebas ---
print(f"factorial(5) = {factorial(5)}")              # 120 (5*4*3*2*1)
print(f"factorial(0) = {factorial(0)}")              # 1 (caso base)
print(f"factorial(-3) = {factorial(-3)}")            # Error (negativo)
print(f"factorial_iterativo(5) = {factorial_iterativo(5)}")    # 120 (mismo resultado)
print(f"factorial_iterativo(10) = {factorial_iterativo(10)}")  # 3628800
```

## Conceptos: Funciones de Primera Clase

En Python, las funciones son **objetos de primera clase** (first-class objects). Esto significa que una funcion:

1. Se puede **guardar en una variable:** `f = len`
2. Se puede **pasar como argumento:** `aplicar(len, lista)`
3. Se puede **devolver desde otra funcion**

Las funciones que reciben o devuelven otras funciones se llaman **higher-order functions** (funciones de orden superior).

Este concepto es la base de la **programacion funcional** y se usa muchisimo en data science:

- `df.apply(funcion)` -> aplica funcion a cada fila/columna de un DataFrame
- `df.map(funcion)` -> aplica funcion a cada elemento
- `sorted(lista, key=funcion)` -> ordena usando funcion como criterio

```python
# --- Funciones de primera clase: aplicar funcion a cada elemento ---

# CONCEPTO: En Python, las funciones son "objetos de primera clase" (first-class objects).
# Esto significa que puedes guardarlas en variables, pasarlas como argumentos
# y devolverlas desde otras funciones, igual que cualquier otro dato.
def aplicar(func, lista):
    """Aplica una funcion a cada elemento de una lista."""
    resultado = []  # Lista vacia donde guardaremos los resultados
    for elemento in lista:
        # Llamamos a func(elemento) - func es la funcion que nos pasaron como argumento.
        # Esto funciona con CUALQUIER funcion: lambda, def, o incluso funciones nativas como len.
        resultado.append(func(elemento))
    return resultado

duplicados = aplicar(lambda x: x * 2, numeros)
print(f"Duplicar {numeros}: {duplicados}")  # [2, 4, 6, 8, 10]

# --- Elevar al cuadrado ---
cuadrados = aplicar(lambda x: x ** 2, numeros)
print(f"Cuadrados {numeros}: {cuadrados}")  # [1, 4, 9, 16, 25]

# --- Longitud de cada string ---

longitudes = aplicar(len, palabras)
```

```python
# --- BONUS: Equivalencia con map() y filter() ---

numeros = [1, 2, 3, 4, 5]
palabras = ["hola", "mundo", "Python", "es", "genial"]

# CONCEPTO: map(func, iterable) aplica func a cada elemento, igual que nuestra funcion aplicar().
# map() devuelve un iterador (lazy), por eso lo envolvemos con list() para obtener una lista.
# Iterador = no calcula todo de golpe, sino uno a uno cuando se necesita (ahorra memoria).
duplicados_map = list(map(lambda x: x * 2, numeros))

pares = list(filter(lambda x: x % 2 == 0, numeros))
```

## Errores Comunes (evita estos!)

| Error | Ejemplo incorrecto | Correccion |
|:---|:---|:---|
| Olvidar `return` (la funcion devuelve `None`) | `def suma(a, b): a + b` | Agrega `return`: `def suma(a, b): return a + b` |
| Confundir parametros y argumentos | "Los argumentos de la funcion son a y b" | Parametros = en la definicion; argumentos = en la llamada |
| Argumento por defecto mutable | `def f(lista=[]):` (la lista se comparte entre llamadas) | Usa `None`: `def f(lista=None): lista = lista or []` |
| Scope incorrecto (local vs global) | Modificar variable global sin `global` | Pasa la variable como parametro y devuelve con `return` |
| Olvidar llamar a la funcion | `resultado = mi_funcion` (sin parentesis) | Agrega parentesis: `resultado = mi_funcion()` |
