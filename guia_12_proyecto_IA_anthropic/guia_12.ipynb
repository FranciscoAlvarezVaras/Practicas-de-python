```python
# --- Clase ChatBot: clasificacion de intenciones por reglas ---
# CONCEPTO: Este patron se llama "chatbot basado en reglas" o "rule-based chatbot".
# Antes de la IA generativa, TODOS los chatbots funcionaban asi:
# 1. El usuario escribe un mensaje
# 2. El bot busca palabras clave para detectar la "intencion"
# 3. Segun la intencion, elige una respuesta predefinida
class ChatBot:
    def __init__(self, nombre):
        self.nombre = nombre  # nombre del bot para personalizar respuestas
        self.historial = []  # lista que guarda cada interaccion para analisis posterior

        # --- Reglas de clasificacion: mapeo intencion -> palabras clave ---
        # CONCEPTO: Cada intencion tiene una lista de palabras/frases que la activan.
        # Si el mensaje del usuario contiene alguna de estas palabras, se clasifica
        # con esa intencion. Es una forma simple pero efectiva de NLU (Natural Language Understanding).
        self.reglas = {
            "saludo": ["hola", "hey", "buenos dias", "buenas", "hi"],
            "despedida": ["adios", "chao", "hasta luego", "bye", "nos vemos"],
            "estado": ["como estas", "que tal", "como te va"],
            "nombre": ["como te llamas", "tu nombre", "quien eres"],
            "ayuda": ["ayuda", "help", "necesito", "puedes"],
            "python": ["python", "programar", "codigo", "programa"],
            "gracias": ["gracias", "thanks", "genial", "perfecto"],
        }

        # --- Respuestas por intencion: multiples opciones por variedad ---
        # CONCEPTO: Tener varias respuestas por intencion y elegir una al azar
        # hace que el bot parezca menos robotico. Es un truco clasico.
        self.respuestas = {
            "saludo": ["Hola! En que puedo ayudarte?",
                       "Hey! Que necesitas?",
                       "Buenos dias! Estoy aqui para ayudar."],
            "despedida": ["Hasta luego! Fue un placer.",
                          "Adios! Vuelve cuando quieras."],
            "estado": ["Estoy funcionando perfectamente! Y tu?",
                       "Muy bien, gracias por preguntar!"],
            # f-string dentro de la lista para incluir el nombre del bot
            "nombre": [f"Me llamo {nombre}. Soy un chatbot creado en Python!"],
            "ayuda": ["Claro! Puedo hablar sobre Python, responder preguntas "
                      "basicas, o simplemente charlar."],
            "python": ["Python es un lenguaje genial! Que quieres saber?",
                       "Me encanta Python! Es muy versatil."],
            "gracias": ["De nada! Para eso estoy.",
                        "Un placer! Algo mas?"],
            # "default" es el fallback: cuando no se detecta ninguna intencion
            "default": ["No estoy seguro de como responder a eso...",
                        "Interesante! Podrias ser mas especifico?",
                        "No entiendo bien. Prueba de otra forma."],
        }

    def clasificar(self, mensaje):
        """Detecta la intencion del mensaje."""
        # Normalizamos a minusculas y quitamos espacios para comparar mejor
        mensaje_lower = mensaje.lower().strip()

        # Recorremos cada intencion y sus palabras clave
        for intencion, palabras_clave in self.reglas.items():
            for palabra in palabras_clave:
                # CONCEPTO: "in" busca si la subcadena existe dentro del mensaje
                # "hola" in "hola que tal" -> True
                # Esto permite detectar frases como "como estas" dentro de textos mas largos
                if palabra in mensaje_lower:
                    return intencion  # devolvemos la primera intencion encontrada
        return "default"  # si ninguna regla coincide, devolvemos "default"

    def responder(self, mensaje):
        """Genera una respuesta basada en la intencion detectada."""
        intencion = self.clasificar(mensaje)  # primero clasificamos
        # random.choice() elige una respuesta al azar entre las disponibles
        # para esa intencion, anadiendo variedad al bot
        respuesta = random.choice(self.respuestas[intencion])

        # --- Guardar en historial ---
        # CONCEPTO: Registrar cada interaccion es fundamental para:
        # 1. Analizar que preguntan los usuarios
        # 2. Detectar intenciones que fallan mucho (mejorar reglas)
        # 3. Medir la satisfaccion del usuario
        self.historial.append({
            "usuario": mensaje,
            "bot": respuesta,
            "intencion": intencion,
            # strftime formatea la fecha/hora como string legible
            "timestamp": datetime.now().strftime("%H:%M:%S")
        })

        return respuesta
```

El analisis de sentimiento cuenta palabras positivas y negativas para determinar la polaridad. La extraccion de entidades busca patrones simples (@ para emails, `isdigit()` para numeros, mayusculas para nombres). El resumen puntua oraciones por cantidad de palabras significativas (no stop words) y devuelve las mejores.

```python
# --- Vocabularios para analisis de sentimiento ---
# CONCEPTO: Usamos set() en vez de list() porque la busqueda en sets es O(1)
# (tiempo constante), mientras que en listas es O(n) (tiempo lineal).
# Como vamos a buscar muchas palabras, sets son MUCHO mas rapidos.
PALABRAS_POSITIVAS = {"bueno", "genial", "excelente", "fantastico", "increible",
                      "perfecto", "maravilloso", "feliz", "amor", "mejor",
                      "bien", "bonito", "facil", "rapido", "util"}

PALABRAS_NEGATIVAS = {"malo", "terrible", "horrible", "odio", "peor", "feo",
                      "triste", "error", "problema", "dificil", "lento",
                      "complicado", "aburrido", "confuso", "frustrado"}

# CONCEPTO: Stop words = palabras funcionales sin significado propio.
# "el", "de", "en" no aportan informacion sobre el contenido del texto.
# Filtrarlas mejora el analisis porque nos quedamos solo con palabras significativas.
STOP_WORDS = {"el", "la", "los", "las", "un", "una", "de", "del", "en",
              "que", "es", "por", "con", "se", "su", "para", "como", "al",
              "no", "si", "ya", "o", "y", "a", "e", "muy", "mas", "pero"}


# --- Funcion: analisis de sentimiento basico ---
def analizar_sentimiento(texto):
    """Analiza si un texto es positivo, negativo o neutro."""
    # Convertimos todo a minusculas para comparar sin distinguir mayusculas
    palabras = texto.lower().split()
    # strip(".,!?;:") quita signos de puntuacion de los extremos de cada palabra
    # Sin esto, "genial!" no coincidiria con "genial" en el vocabulario
    palabras = [p.strip(".,!?;:") for p in palabras]

    # CONCEPTO: Conteo de sentimiento - sumamos cuantas palabras de cada tipo hay.
    # sum(1 for ...) es una expresion generadora que cuenta coincidencias.
    # Es equivalente a un bucle for con contador, pero mas compacto.
    positivas = sum(1 for p in palabras if p in PALABRAS_POSITIVAS)
    negativas = sum(1 for p in palabras if p in PALABRAS_NEGATIVAS)

    # La polaridad se decide por mayoria: mas positivas -> positivo, etc.
    if positivas > negativas:
        return "positivo", positivas, negativas
    elif negativas > positivas:
        return "negativo", positivas, negativas
    else:
        return "neutro", positivas, negativas


# --- Funcion: extraccion de entidades basicas ---
def extraer_entidades(texto):
    """Extrae entidades basicas de un texto."""
    palabras = texto.split()  # dividimos por espacios
    # Inicializamos un diccionario con listas vacias para cada tipo de entidad
    entidades = {"emails": [], "numeros": [], "posibles_nombres": []}

    for palabra in palabras:
        # Limpiamos signos de puntuacion de los extremos
        limpia = palabra.strip(".,!?;:()")

        if "@" in limpia:
            # CONCEPTO: Deteccion de emails - si tiene "@", probablemente es un email.
            # En NLP real se usarian expresiones regulares (regex) mas precisas.
            entidades["emails"].append(limpia)
        elif limpia.isdigit():
            # isdigit() devuelve True si TODOS los caracteres son digitos
            # Convertimos a int para tener el valor numerico, no el string
            entidades["numeros"].append(int(limpia))
        elif limpia[0:1].isupper() and len(limpia) > 1 and limpia.isalpha():
            # CONCEPTO: Heuristica para nombres propios:
            # - limpia[0:1].isupper(): la primera letra es mayuscula
            #   (usamos [0:1] en vez de [0] para evitar IndexError si esta vacio)
            # - len > 1: descartamos letras sueltas como "A" o "Y"
            # - isalpha(): solo letras (no numeros ni simbolos)
            entidades["posibles_nombres"].append(limpia)

    return entidades


# --- Funcion: resumen extractivo de texto ---
def resumir(texto, n_oraciones=2):
    """Devuelve las n oraciones mas importantes."""
    # Dividimos el texto en oraciones usando el punto como separador
    # strip() quita espacios y filtramos las vacias con if o.strip()
    oraciones = [o.strip() for o in texto.split(".") if o.strip()]

    # --- Puntuacion de cada oracion ---
    # CONCEPTO: Resumen extractivo = seleccionar las oraciones mas importantes
    # del texto original (a diferencia del resumen abstractivo que genera texto nuevo).
    # Puntuamos cada oracion por la cantidad de "palabras significativas"
    # (excluyendo stop words). Mas palabras significativas = mas informativa.
    puntuaciones = []
    for oracion in oraciones:
        palabras = oracion.lower().split()
        # Filtramos stop words para quedarnos solo con palabras con contenido
        significativas = [p for p in palabras if p not in STOP_WORDS]
        puntuaciones.append((oracion, len(significativas)))

    # Ordenamos por puntuacion (descendente) y tomamos las mejores n
    ordenadas = sorted(puntuaciones, key=lambda x: x[1], reverse=True)
    # Anadimos "." al final de cada oracion porque se perdio al hacer split(".")
    return [o[0] + "." for o in ordenadas[:n_oraciones]]


# --- Pruebas ---

print("\nExtraccion de entidades:")

texto_largo = ("Python es un lenguaje de programacion poderoso. "

for oracion in resumen:
    print(f"  {oracion}")
```

El sistema Q&A funciona comparando las palabras clave de la pregunta con cada oracion del texto. La oracion que comparte mas palabras clave es la respuesta mas probable. Este enfoque se llama "busqueda por similitud de texto" y es la base de muchos sistemas de recuperacion de informacion.

```python
# --- Clase SistemaQA: busqueda por similitud de texto ---
# CONCEPTO: Este es un sistema de "recuperacion de informacion" (Information Retrieval).
# Dado un texto base y una pregunta, busca la oracion del texto que mejor
# responde a la pregunta. Es la base de como funcionan los buscadores y
# los sistemas RAG (Retrieval-Augmented Generation) de IA moderna.
class SistemaQA:
    def __init__(self, texto):
        self.texto = texto
        # Dividimos el texto en oraciones al instanciar la clase
        # Anadimos "." al final de cada oracion porque split(".") lo elimina
        # Filtramos oraciones vacias con if o.strip()
        self.oraciones = [o.strip() + "." for o in texto.split(".")
                          if o.strip()]

    def _palabras_clave(self, pregunta):
        """Extrae palabras clave de una pregunta."""
        # CONCEPTO: El prefijo _ indica que es un metodo "privado" (convencion).
        # No es obligatorio, pero indica que no deberia llamarse desde fuera.
        # Normalizamos a minusculas y quitamos signos de pregunta
        palabras = pregunta.lower().strip("?!").split()
        # Filtramos: quitamos stop words y palabras cortas (<=2 caracteres)
        # Las palabras cortas como "de", "en", "el" no ayudan a buscar
        return [p for p in palabras if p not in STOP_WORDS and len(p) > 2]

    def responder(self, pregunta):
        """Busca la oracion mas relevante para la pregunta."""
        claves = self._palabras_clave(pregunta)

        if not claves:
            return "No entiendo la pregunta."

        mejor_oracion = ""
        mejor_puntuacion = 0

        # --- Busqueda por similitud: comparar pregunta con cada oracion ---
        for oracion in self.oraciones:

            puntuacion = sum(1 for c in claves if c in oracion_lower)

            # Patron "buscar el mejor": actualizamos si encontramos algo mejor
            if puntuacion > mejor_puntuacion:
                mejor_puntuacion = puntuacion
                mejor_oracion = oracion

        if mejor_puntuacion > 0:
            return mejor_oracion  # devolvemos la oracion mas relevante
        return "No encontre informacion sobre eso en el texto."
```

El Pipeline es un patron muy usado en data engineering. Los datos pasan por una serie de transformaciones en orden, como una cadena de montaje. Cada paso hace una sola cosa (principio de responsabilidad unica). El log permite rastrear que paso a los datos en cada etapa, lo cual es fundamental para debugging.

```python
# --- Clase Pipeline: patron ETL (Extract, Transform, Load) ---
# CONCEPTO: ETL es el proceso fundamental en data engineering:
#   E (Extract) = obtener datos crudos de alguna fuente
#   T (Transform) = limpiar, validar, enriquecer los datos
#   L (Load) = guardar los datos procesados en su destino
# El patron Pipeline encadena transformaciones en orden, como una cadena de montaje.
# Cada paso hace UNA sola cosa (principio de responsabilidad unica).
class Pipeline:
    def __init__(self, nombre="Pipeline"):
        self.nombre = nombre
        self.pasos = []  # lista de tuplas (funcion, descripcion)
        self.log = []  # registro de lo que ocurre en cada ejecucion

    def agregar_paso(self, funcion, descripcion=""):
        """Agrega un paso al pipeline."""
        # CONCEPTO: Las funciones en Python son "objetos de primera clase",
        # lo que significa que se pueden pasar como argumento a otras funciones,
        # guardar en listas, etc. Aqui guardamos la funcion para ejecutarla despues.
        self.pasos.append((funcion, descripcion))

    def ejecutar(self, datos):
        """Ejecuta todos los pasos en orden."""
        self.log = []  # reseteamos el log de ejecuciones anteriores
        self.log.append(f"[INICIO] {self.nombre} - {len(datos)} registros")

        # CONCEPTO: enumerate() nos da el indice i (para numerar los pasos)
        # y el par (funcion, descripcion) en cada iteracion.
        for i, (funcion, descripcion) in enumerate(self.pasos):
            antes = len(datos)  # cuantos registros habia antes del paso
            datos = funcion(datos)  # ejecutamos la transformacion
            # isinstance comprueba si datos es una lista (para contar registros)
            despues = len(datos) if isinstance(datos, list) else "N/A"
            msg = f"[PASO {i+1}] {descripcion}: {antes} -> {despues} registros"
            self.log.append(msg)
            print(f"  {msg}")  # mostramos el progreso en tiempo real

        self.log.append(f"[FIN] Pipeline completado")
        return datos


# --- Funciones de transformacion (cada una hace UNA cosa) ---

def normalizar_nombres(datos):
    """Normaliza nombres a formato Titulo."""
    for d in datos:
        # strip() quita espacios, title() pone primera letra en mayuscula
        d["nombre"] = d["nombre"].strip().title()
        d["apellido"] = d["apellido"].strip().title()
    return datos  # siempre devolvemos los datos para encadenar


def validar_emails(datos):
    """Marca emails validos/invalidos."""
    for d in datos:
        email = d.get("email", "")  # .get() evita KeyError si no existe
        # Un email basico debe tener "@" y "." para ser valido
        if "@" not in email or "." not in email:
            d["email_valido"] = False
        else:
            d["email_valido"] = True
    return datos


def agregar_rango_edad(datos):
    """Clasifica empleados por rango de edad."""
    for d in datos:
        edad = d.get("edad", 0)
        # Clasificacion en 3 rangos usando if/elif/else
        if edad < 30:
            d["rango_edad"] = "joven"
        elif edad < 45:
            d["rango_edad"] = "adulto"
        else:
            d["rango_edad"] = "senior"
    return datos


def agregar_timestamp(datos):
    """Agrega fecha/hora de procesamiento a cada registro."""
    # CONCEPTO: Registrar CUANDO se procesaron los datos es fundamental
    # para trazabilidad y auditorias. strftime formatea la fecha como string.
    ts = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    for d in datos:
        d["procesado_en"] = ts  # mismo timestamp para todo el lote
    return datos


def eliminar_duplicados_pipeline(datos):
    """Elimina registros duplicados por email."""
    # Mismo patron que en guia_11: set para detectar duplicados en O(1)
    vistos = set()
    unicos = []
    for d in datos:
        clave = d.get("email", "")  # usamos email como identificador unico
        if clave not in vistos:
            vistos.add(clave)
            unicos.append(d)
    return unicos


# --- Crear y ejecutar el pipeline ---

# E (Extract): obtenemos datos crudos del generador
gen = GeneradorDatos()
datos_crudos = gen.dataset(20)  # 20 registros de prueba

# T (Transform): configuramos los pasos de transformacion
etl = Pipeline("ETL Empleados")
etl.agregar_paso(normalizar_nombres, "Normalizar nombres")
etl.agregar_paso(validar_emails, "Validar emails")
etl.agregar_paso(agregar_rango_edad, "Agregar rango de edad")
etl.agregar_paso(agregar_timestamp, "Agregar timestamp")
etl.agregar_paso(eliminar_duplicados_pipeline, "Eliminar duplicados")

# Ejecutamos el pipeline (los datos pasan por todos los pasos en orden)
datos_procesados = etl.ejecutar(datos_crudos)

# L (Load): mostramos resultados (en produccion se guardarian en BD o archivo)
print(f"\nResultado: {len(datos_procesados)} registros procesados")
# json.dumps convierte un diccionario a string JSON formateado para mostrarlo bonito
print(f"Ejemplo: {json.dumps(datos_procesados[0], indent=4, ensure_ascii=False)}")
```

La clase `AsistentePython` intenta conectarse a la API de Anthropic. Si no esta disponible (sin libreria o sin API key), usa un sistema de respuestas simuladas como fallback.
```python
# --- Clase AsistentePython: integracion con la API de Claude ---
# CONCEPTO: Esta clase demuestra el "patron fallback" (plan B):
# intenta usar la API real, pero si no esta disponible, usa respuestas simuladas.
# Esto es MUY comun en produccion: siempre tener una alternativa si el servicio falla.
class AsistentePython:
    """Asistente de Python usando la API de Claude.
    NOTA: Cada llamada a la API tiene un coste. Revisa tus limites en
    console.anthropic.com antes de hacer muchas consultas."""

    def __init__(self):
        self.historial = []  # guardamos todas las interacciones
        self.api_disponible = False  # flag que indica si la API funciona

        # --- Intento de conexion con la API ---
        # CONCEPTO: try/except con multiples excepciones para manejar
        # diferentes tipos de fallos al inicializar la API.
        try:
            import anthropic  # intenta importar la libreria
            # Anthropic() busca automaticamente la variable de entorno ANTHROPIC_API_KEY
            # No necesitas pasar la key como argumento si la variable esta configurada
            self.client = anthropic.Anthropic()
            self.api_disponible = True
            print("API de Anthropic conectada!")
        except ImportError:
            # Si la libreria no esta instalada (falta pip install anthropic)
            print("[INFO] anthropic no instalado. pip install anthropic")
        except Exception as e:
            # Cualquier otro error (API key invalida, sin conexion, etc.)
            print(f"[INFO] API no disponible: {e}")

    def preguntar(self, pregunta, modo_profesor=False):
        """Envia una pregunta a Claude y devuelve la respuesta."""
        # Si la API no esta disponible, usamos el fallback (respuestas simuladas)
        if not self.api_disponible:
            return self._respuesta_simulada(pregunta)

        # --- Construir el "system prompt" segun el modo ---
        # CONCEPTO: El "system prompt" es una instruccion que le dice al modelo
        # COMO debe comportarse. Es como darle un "rol" al asistente.
        # Cambiar el system prompt cambia completamente el estilo de respuesta.
        if modo_profesor:
            sistema = ("Eres un profesor de Python del MIT. Explica los "
                       "conceptos paso a paso, con ejemplos simples. "
                       "Usa analogias para que un novato entienda.")
        else:
            sistema = ("Eres un asistente de Python. Responde de forma "
                       "concisa y con ejemplos de codigo cuando sea util.")

        try:
            # --- Llamada a la API de Claude ---
            # CONCEPTO: client.messages.create() es una llamada HTTP POST a la API.
            # Es sincrona: el codigo espera hasta recibir la respuesta completa.
            message = self.client.messages.create(
                model="claude-sonnet-4-5-20250929",  # modelo a usar
                max_tokens=1024,  # limite de tokens en la respuesta (1 token ~ 0.75 palabras)
                system=sistema,  # instrucciones de comportamiento
                # messages es una lista de turnos de conversacion
                # "role": "user" indica que es un mensaje del usuario
                messages=[{"role": "user", "content": pregunta}]
            )
            # La respuesta viene en message.content, que es una lista de bloques.
            # [0].text accede al texto del primer bloque (normalmente solo hay uno).
            respuesta = message.content[0].text
        except Exception as e:
            # Si la llamada falla (timeout, error de red, limite de uso, etc.)
            respuesta = f"Error al consultar la API: {e}"

        # --- Guardar en historial ---
        # Registramos cada interaccion para poder analizarla o exportarla despues
        self.historial.append({
            "pregunta": pregunta,
            "respuesta": respuesta,
            "modo": "profesor" if modo_profesor else "asistente",
            "timestamp": datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        })

        return respuesta

    def _respuesta_simulada(self, pregunta):
        """Respuesta simulada cuando no hay API disponible."""
        # CONCEPTO: Patron fallback - diccionario de respuestas predefinidas.
        # Buscamos si alguna palabra clave de las respuestas aparece en la pregunta.
        # Es el mismo enfoque del ChatBot del ejercicio 1, pero mas simple.
        respuestas = {
            "lista": "Una lista en Python es una coleccion ordenada y "
                     "modificable. Se crea con []. Ejemplo: mi_lista = [1, 2, 3]",
            "funcion": "Una funcion se define con 'def'. Ejemplo:\n"
                       "def saludar(nombre):\n    return f'Hola {nombre}'",
            "diccionario": "Un diccionario almacena pares clave:valor. "
                           "Ejemplo: persona = {'nombre': 'Ana', 'edad': 25}",
            "for": "El bucle for recorre elementos: for x in [1,2,3]: print(x)",
            "clase": "Una clase agrupa datos y funciones:\n"
                     "class Perro:\n    def __init__(self, nombre):\n"
                     "        self.nombre = nombre",
        }

        pregunta_lower = pregunta.lower()
        for clave, resp in respuestas.items():
            if clave in pregunta_lower:  # buscamos la palabra clave en la pregunta
                return f"[SIMULADO] {resp}"

        # Si ninguna palabra clave coincide, devolvemos un mensaje generico
        return ("[SIMULADO] Para obtener respuestas reales, configura "
                "tu API key de Anthropic. Visita: https://console.anthropic.com")

    def guardar_historial(self, archivo="historial_chat.json"):
        """Guarda el historial en JSON."""
        # CONCEPTO: Persistencia de datos - guardamos el historial en un archivo
        # JSON para poder consultarlo despues o analizarlo.
        with open(archivo, "w", encoding="utf-8") as f:
            json.dump(self.historial, f, indent=4, ensure_ascii=False)
```

> **Recuerda:** Como funciona la API de Claude:

| Paso | Codigo | Descripcion |
|------|--------|-------------|
| 1. Instalar | `pip install anthropic` | Instalar la libreria cliente |
| 2. Configurar | `export ANTHROPIC_API_KEY="..."` | Configurar la variable de entorno |
| 3. Crear cliente | `client = anthropic.Anthropic()` | Se conecta usando la API key |
| 4. Enviar mensaje | `client.messages.create(...)` | Enviar modelo, tokens, mensajes |
| 5. Leer respuesta | `message.content[0].text` | Texto generado por el modelo |

Es el mismo patron que usamos con `requests.get()` en la Guia 9: envias datos, recibes respuesta, la procesas. La diferencia es que aqui la respuesta es **inteligente**.

## Errores Comunes

| Error | Problema | Solucion |
|---|---|---|
| No manejar timeouts de la API | El programa se cuelga esperando respuesta | Usar `timeout` en las llamadas y `try/except` |
| Exponer la API key en el codigo | Riesgo de seguridad (robo de credenciales) | Usar variables de entorno: `os.environ["ANTHROPIC_API_KEY"]` |
| No tener plan B (fallback) | La app no funciona si la API esta caida | Implementar respuestas simuladas como fallback |
| No respetar rate limits | Error 429 (Too Many Requests) | Implementar reintentos con espera exponencial |
| No validar la respuesta de la API | Crash si la respuesta viene en formato inesperado | Verificar `message.content` antes de acceder a `.text` |
