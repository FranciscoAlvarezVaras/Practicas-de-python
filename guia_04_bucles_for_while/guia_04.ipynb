## Conceptos Clave: El Bucle `for`

El bucle `for` recorre cada elemento de una secuencia (lista, string, range) **uno a uno**.

```python
for elemento in secuencia:
    # codigo que se ejecuta en cada vuelta
```

**Ejemplo:**
```python
for fruta in ["manzana", "pera", "uva"]:
    print(fruta)
```

La funcion `range()` genera secuencias de numeros:

| Forma | Resultado | Descripcion |
|-------|-----------|-------------|
| `range(5)` | 0, 1, 2, 3, 4 | Desde 0 hasta 4 |
| `range(2, 7)` | 2, 3, 4, 5, 6 | Desde 2 hasta 6 |
| `range(0, 10, 2)` | 0, 2, 4, 6, 8 | De 2 en 2 |
| `range(10, 0, -1)` | 10, 9, 8, ..., 1 | Cuenta atras |

> **Recuerda:** `range(fin)` **nunca** incluye el valor `fin`.

## Conceptos Clave: Bucles Anidados

Un **bucle anidado** es un bucle dentro de otro. El bucle interior se ejecuta **completamente** en cada vuelta del bucle exterior.

```python
for fila in range(3):
    for columna in range(4):
        print("*", end=" ")
    print()  # salto de linea al final de cada fila
```

> **Tip:** `print(x, end=" ")` imprime **sin** saltar de linea (agrega un espacio al final). `print()` imprime un salto de linea vacio.

## Conceptos Clave: `enumerate()`

Cuando necesitas el **indice** y el **valor** al mismo tiempo, usa `enumerate()` en lugar de `range(len(...))`:

```python
# Forma clasica (menos pythonica)
for i in range(len(frase)):
    print(i, frase[i])

# Forma recomendada (pythonica)
for indice, letra in enumerate(frase):
    print(indice, letra)
```

> **Recuerda:** `enumerate()` devuelve pares `(indice, valor)` automaticamente. Es mas limpio y menos propenso a errores.

## Conceptos Clave: El Bucle `while`

El bucle `while` repite un bloque de codigo **mientras** una condicion sea `True`.

```python
while condicion:
    # codigo
    # actualizar algo (si no, bucle infinito!)
```

**`for` vs `while`:**

| Usar `for` cuando... | Usar `while` cuando... |
|----------------------|------------------------|
| Sabes cuantas vueltas dar | No sabes cuantas vueltas seran |
| Recorres una lista o range | Dependes de una condicion |
| Ejemplo: recorrer alumnos | Ejemplo: pedir input hasta que sea valido |

> **Ojo:** si la condicion nunca se hace `False`, el bucle es **infinito**. Siempre asegurate de que algo cambie para que el `while` termine.

2. Inicializa una variable `total` en 0
3. Con un `while`, suma los numeros uno a uno
4. Imprime cada suma parcial
5. Cuando `total` pase de 100, imprime `"Limite superado!"` y detente

> **Tip:** necesitas un indice que avance manualmente (`i += 1`).

### Tu turno

````python
# TU CODIGO AQUI

````

<details>
<summary><b>Ver solucion - Ejercicio 9</b></summary>

En un `while`, nosotros controlamos el indice manualmente con `i += 1`. La condicion del `while` se comprueba **antes** de cada vuelta. Usamos `break` para salir inmediatamente cuando se supera el limite.

```python
# --- Ejercicio 9: While - suma acumulada hasta superar 100 ---

numeros = [15, 20, 8, 35, 12, 25, 30]  # Lista de numeros a sumar
total = 0  # Acumulador: empieza en 0 y se le van sumando numeros
i = 0  # Indice manual: en un while, nosotros controlamos el avance

# CONCEPTO: El bucle "while" se ejecuta MIENTRAS la condicion sea True
# Aqui tenemos DOS condiciones unidas con "and":
# 1) i < len(numeros) -> no pasarnos del final de la lista (evita IndexError)
# 2) total <= 100 -> seguir mientras no superemos el limite
# CONCEPTO: Python evalua de izquierda a derecha con short-circuit:
# si i >= len(numeros), NO evalua total <= 100 (evita error)
while i < len(numeros) and total <= 100:
    total += numeros[i]  # Sumamos el numero en la posicion actual al total
    print(f"Sumando {numeros[i]} -> Total: {total}")

    # Comprobamos si acabamos de superar el limite de 100
    if total > 100:
        print("Limite superado!")
        # CONCEPTO: "break" sale del bucle INMEDIATAMENTE
        # Sin break, el while volveria a evaluar la condicion y saldria igual,
        # pero break es mas explicito sobre nuestra intencion
        break

    # CONCEPTO: En un while, debemos incrementar el indice MANUALMENTE
    # Si olvidamos i += 1, el bucle seria INFINITO (siempre procesaria numeros[0])
    i += 1
```

</details>

## Conceptos Clave: `break` y `continue`

Dos instrucciones para controlar el flujo dentro de un bucle:

- **`break`**: sale del bucle **inmediatamente** (no ejecuta mas vueltas)
- **`continue`**: salta al **siguiente paso** del bucle (no ejecuta el resto del bloque en esa vuelta)

```python
for i in range(5):
    if i == 2:
        continue  # salta el 2
    print(i)
# Salida: 0, 1, 3, 4 (se salta el 2)
```

> **Recuerda:** `break` termina el bucle entero; `continue` solo salta la vuelta actual.

## Conceptos Clave: Enumeracion Exhaustiva

La **enumeracion exhaustiva** ("exhaustive enumeration", MIT 6.0001) consiste en probar **todos** los posibles candidatos sistematicamente. Es un enfoque de "fuerza bruta" pero funciona bien para rangos pequenos.

**Ejemplo clasico:** para saber si un numero `n` es primo, probamos todos los divisores posibles desde 2 hasta la raiz cuadrada de `n`.

**Por que solo hasta la raiz cuadrada?**
- Si `n = 36`, sus divisores son: 1, 2, 3, 4, 6, 9, 12, 18, 36
- `sqrt(36) = 6`. Los divisores vienen en **parejas**: 2x18, 3x12, 4x9, 6x6
- Si no encontramos divisor hasta `sqrt(n)`, no lo hay despues tampoco
- Esto hace el algoritmo **mucho mas eficiente**

> **Recuerda:** `int(25 ** 0.5)` = `int(5.0)` = `5`; `int(24 ** 0.5)` = `int(4.898...)` = `4`.

## Ejercicio 12 de 13: Numeros primos | Dificil

> **Conceptos clave:** enumeracion exhaustiva | `for` anidado | `break` | raiz cuadrada | flag booleano

Escribe codigo que encuentre todos los **numeros primos entre 2 y 100**.

Un numero es primo si solo es divisible por 1 y por si mismo.

Para cada primo encontrado, imprimelo. Al final, imprime el total de primos encontrados.

> **Tip:** para cada numero `n`, comprueba si algun numero desde 2 hasta `int(n ** 0.5) + 1` lo divide exactamente. Si ninguno lo divide, es primo.

**Ejemplo de salida:**
```
2 es primo
3 es primo
5 es primo
...
Total de primos entre 2 y 100: XX
```

### Tu turno

````python
# TU CODIGO AQUI

````

<details>
<summary><b>Ver solucion - Ejercicio 12</b></summary>

Usamos **enumeracion exhaustiva**: para cada `n`, probamos todos los divisores de 2 a `sqrt(n)`. Si encontramos uno, marcamos `es_primo = False` y usamos `break` para no seguir comprobando (optimizacion: ahorra tiempo en numeros grandes).

```python
# --- Ejercicio 12: Encontrar numeros primos entre 2 y 100 ---

contador_primos = 0  # Acumulador para contar el total de primos encontrados

# CONCEPTO: range(2, 101) va de 2 a 100. Empezamos en 2 porque 0 y 1 NO son primos
# Un numero primo solo es divisible por 1 y por si mismo
for n in range(2, 101):
    # --- Flag booleano: asumimos que es primo hasta demostrar lo contrario ---
    # CONCEPTO: Patron de "flag" (bandera): se inicializa en True y se cambia a False
    # si encontramos evidencia en contra. Es un patron muy comun en algoritmos
    es_primo = True

    # --- Enumeracion exhaustiva: probar todos los posibles divisores ---
    # CONCEPTO: Solo probamos divisores de 2 hasta raiz cuadrada de n
    # Porque si n = a * b, uno de los dos (a o b) es <= sqrt(n)
    # Ejemplo: 36 = 6*6 = 4*9 = 3*12 -> basta probar hasta 6 (sqrt(36))
    # int(n ** 0.5) calcula la raiz cuadrada entera: n ** 0.5 = sqrt(n)
    # +1 porque range no incluye el fin
    for divisor in range(2, int(n ** 0.5) + 1):
        if n % divisor == 0:  # Si el resto es 0, n es divisible -> NO es primo
            es_primo = False
            # CONCEPTO: "break" optimiza: si ya encontramos un divisor, no
            # necesitamos probar mas. Salimos del bucle interior inmediatamente
            break

    # Si despues de probar todos los divisores la bandera sigue en True -> es primo
    if es_primo:
        print(f"  {n} es primo")
        contador_primos += 1

print(f"\nTotal de primos entre 2 y 100: {contador_primos}")
```

## Conceptos Clave: Patrones con Bucles y Multiplicacion de Strings

En Python, puedes **multiplicar** strings para repetirlos:

```python
" " * 3    # "   " (3 espacios)
"*" * 5    # "*****" (5 estrellas)
```

La **concatenacion** `+` los une:
```python
"   " + "*****"  # "   *****"
```

> **Recuerda:** el truco para construir patrones visuales esta en encontrar la **formula matematica** que relaciona la fila (`i`) con la cantidad de caracteres.

**Formulas tipicas para un diamante de tamanio `n`:**
- `estrellas = 2 * i + 1` (genera impares: 1, 3, 5, 7, 9)
- `espacios = n - 1 - i` (va decreciendo para centrar las estrellas)

## Ejercicio 13 de 13: Patron de diamante | Dificil

> **Conceptos clave:** bucles anidados | multiplicacion de strings | formulas matematicas | `range()` negativo

Dado `n = 5`, imprime un patron de diamante usando `*` y espacios:

```
    *
   ***
  *****
 *******
*********
 *******
  *****
   ***
    *
```

> **Tip:** para cada fila, calcula cuantos espacios poner antes de las estrellas y cuantas estrellas poner. Usa `" " * espacios + "*" * estrellas` para construir cada linea.

**Formulas:**
- Parte superior: para `i` en `range(n)`, `estrellas = 2*i + 1`, `espacios = n - 1 - i`
- Parte inferior: para `i` en `range(n-2, -1, -1)`, misma formula

### Tu turno

````python
# TU CODIGO AQUI

````

<details>
<summary><b>Ver solucion - Ejercicio 13</b></summary>

Desglose paso a paso para `n = 5`:

**Parte superior** - `range(5)` --> `i = 0, 1, 2, 3, 4`

| `i` | `espacios` | `estrellas` | Salida |
|-----|-----------|------------|--------|
| 0 | 4 | 1 | `    *` |
| 1 | 3 | 3 | `   ***` |
| 2 | 2 | 5 | `  *****` |
| 3 | 1 | 7 | ` *******` |
| 4 | 0 | 9 | `*********` |

**Parte inferior** - `range(3, -1, -1)` --> `i = 3, 2, 1, 0` (espejo de la superior, sin repetir el centro)

```python
# --- Ejercicio 13: Patron de diamante con n=5 ---

n = 5  # Tamanio del diamante (determina la fila central con 2*n-1 = 9 estrellas)

# --- Parte superior (incluyendo la fila central) ---
# CONCEPTO: range(n) genera i = 0, 1, 2, 3, 4 (5 filas)
for i in range(n):
    # CONCEPTO: Formulas matematicas que relacionan la fila (i) con el patron:
    # espacios = n - 1 - i  -> decrece conforme bajamos (4, 3, 2, 1, 0)
    # estrellas = 2 * i + 1 -> genera impares: 1, 3, 5, 7, 9
    # La suma espacios + estrellas = n - 1 - i + 2*i + 1 = n + i (crece, centrando)
    espacios = n - 1 - i
    estrellas = 2 * i + 1
    # CONCEPTO: Multiplicacion de strings -> " " * 3 = "   " (3 espacios)
    # Concatenacion con + une los espacios y las estrellas en una sola linea
    print(" " * espacios + "*" * estrellas)

# --- Parte inferior (espejo de la superior, SIN repetir la fila central) ---
# CONCEPTO: range(n-2, -1, -1) genera i = 3, 2, 1, 0 (4 filas)
# Empezamos en n-2 (no n-1) para NO repetir la fila central
# Terminamos en -1 para INCLUIR i=0 (la punta inferior)
for i in range(n - 2, -1, -1):
    # Mismas formulas que arriba: como i va decreciendo, el patron se invierte
    espacios = n - 1 - i  # Ahora CRECE: 1, 2, 3, 4
    estrellas = 2 * i + 1  # Ahora DECRECE: 7, 5, 3, 1
    print(" " * espacios + "*" * estrellas)
```

## Errores Comunes (evita estos!)

| Error | Ejemplo incorrecto | Correccion |
|:---|:---|:---|
| Bucle `while` infinito | `while True:` sin `break` ni cambio en condicion | Asegurate de incrementar el contador o cambiar la condicion dentro del bucle |
| Modificar lista mientras se itera con `for` | `for x in lista: lista.remove(x)` | Itera sobre una copia: `for x in lista[:]` o usa list comprehension |
| Off-by-one en `range()` | `range(1, 10)` para obtener 1-10 (falta el 10) | Recuerda: `range(1, 11)` incluye del 1 al 10 |
| Confundir `break` y `continue` | Usar `break` cuando quieres saltar una iteracion | `break` = salir del bucle; `continue` = saltar a la siguiente vuelta |
| Indentacion incorrecta en bucles anidados | Codigo del bucle interior al nivel del exterior | Cada nivel de anidamiento necesita un nivel mas de indentacion |
