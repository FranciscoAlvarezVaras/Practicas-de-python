Practicamos los metodos principales de diccionarios: `.keys()`, `.values()`, `.get()`, `.items()`, `.pop()` y `.clear()`.

```python
# --- Ejercicio 2: Metodos principales de diccionarios ---

# CONCEPTO: Un diccionario almacena pares clave:valor
# Las claves deben ser inmutables (str, int, tuple) y unicas
# Los valores pueden ser de cualquier tipo
agenda = {
    "Ana": "654123987",
    "Luis": "678555111",
    "Marta": "622444333"
}

# --- .keys() -> obtener todas las claves ---
# CONCEPTO: .keys() devuelve un "view object", lo envolvemos en list() para verlo como lista
print(f"Claves: {list(agenda.keys())}")

# --- .values() -> obtener todos los valores ---
# CONCEPTO: .values() devuelve todos los valores del diccionario
print(f"Valores: {list(agenda.values())}")

# --- .get() -> acceso SEGURO a una clave ---
# CONCEPTO: .get() NO lanza error si la clave no existe (a diferencia de agenda["clave"])
# Esto evita KeyError y hace el codigo mas robusto
print(f"Luis: {agenda.get('Luis')}")

# CONCEPTO: .get() con clave inexistente devuelve None (sin error)
# Esto es MUY util para evitar crashes cuando no sabes si la clave existe
print(f"Pedro: {agenda.get('Pedro')}")

# CONCEPTO: .get() acepta un segundo argumento como valor por defecto
# Si la clave no existe, devuelve ese valor en vez de None
print(f"Pedro: {agenda.get('Pedro', 'No encontrado')}")

# --- Agregar un nuevo par clave:valor ---
# CONCEPTO: Para agregar o modificar, se usa la sintaxis dict[clave] = valor
# Si la clave ya existe, se SOBREESCRIBE el valor anterior
agenda["Pedro"] = "699888777"

# --- .items() -> obtener pares (clave, valor) ---
# CONCEPTO: .items() devuelve tuplas (clave, valor), ideal para desempaquetar en un for
print("\nTodos los contactos:")
for nombre, telefono in agenda.items():  # Desempaquetado automatico de cada tupla
    print(f"  {nombre}: {telefono}")

# --- .pop() -> eliminar y obtener el valor ---
# CONCEPTO: .pop(clave) elimina la clave Y devuelve su valor
# Es diferente a "del agenda['Marta']" que elimina pero NO devuelve nada
marta_tel = agenda.pop("Marta")
print(f"\nEliminada Marta. Su telefono era: {marta_tel}")
print(f"Agenda actual: {agenda}")

# --- .clear() -> vaciar completamente el diccionario ---
# CONCEPTO: .clear() elimina TODOS los pares, dejando el diccionario vacio {}
agenda.clear()
print(f"Agenda vacia: {agenda}")
# Comprobamos que esta vacio comparando su longitud con 0
print(f"Esta vacia? {len(agenda) == 0}")
```

</details>

> **Recuerda:** Metodos de diccionarios mas utiles
>
> | Metodo | Descripcion |
> |--------|-------------|
> | `.keys()` | Devuelve todas las claves |
> | `.values()` | Devuelve todos los valores |
> | `.items()` | Devuelve pares (clave, valor) |
> | `.get(clave)` | Obtiene valor sin error si no existe |
> | `.pop(clave)` | Elimina y devuelve el valor |
> | `.update(dict2)` | Agrega/actualiza con otro diccionario |
> | `.clear()` | Vacia el diccionario |

> **Tip:** Usa `.get()` en lugar de `dict[clave]` cuando no estes seguro si la clave existe. Con `.get()` obtienes `None` en vez de un `KeyError`.

Recorremos los metodos mas comunes para manipular listas: agregar, insertar, eliminar, ordenar y vaciar.

```python
# --- Ejercicio 3: Metodos de listas ---

# CONCEPTO: Las listas son MUTABLES (se pueden modificar despues de crearlas)
# A diferencia de las tuplas que son inmutables
peliculas = ["Inception", "The Matrix", "Interstellar", "Gladiator"]
print(f"Original: {peliculas}")

# --- .append(x) -> agregar al FINAL ---
# CONCEPTO: .append() siempre agrega al final de la lista (posicion -1)
# Solo acepta UN argumento. Para agregar varios, usa .extend()
peliculas.append("El Padrino")
print(f"Tras append: {peliculas}")

# --- .insert(posicion, elemento) -> insertar en posicion especifica ---
# CONCEPTO: .insert() desplaza los elementos existentes hacia la derecha
# Los indices despues de la posicion 2 se incrementan en 1
peliculas.insert(2, "Memento")
print(f"Tras insert(2): {peliculas}")

# --- .remove(elemento) -> eliminar por VALOR ---
# CONCEPTO: .remove() busca la PRIMERA aparicion del valor y la elimina
# Si el valor no existe, lanza ValueError. Cuidado con duplicados: solo borra el primero
peliculas.remove("Gladiator")
print(f"Tras remove Gladiator: {peliculas}")

# --- .pop() -> eliminar y devolver por POSICION ---
# CONCEPTO: Sin argumento, .pop() elimina y devuelve el ULTIMO elemento
# Devuelve el elemento eliminado, lo que permite guardarlo en una variable
ultima = peliculas.pop()
print(f"pop() devolvio: {ultima}")
print(f"Tras pop: {peliculas}")

# CONCEPTO: .pop(indice) acepta cualquier posicion valida
# peliculas.pop(0)  -> elimina y devuelve el primero
# peliculas.pop(2)  -> elimina y devuelve el de posicion 2
# Diferencia con .remove(): pop usa INDICE, remove usa VALOR

# --- .sort() -> ordenar la lista IN-PLACE ---
# CONCEPTO: .sort() modifica la lista original (no crea una nueva)
# Para obtener una copia ordenada sin modificar la original, usa sorted(lista)
peliculas.sort()
print(f"Ordenada: {peliculas}")

# --- .clear() -> vaciar la lista ---
# CONCEPTO: .clear() elimina todos los elementos, dejando la lista vacia []
# La variable sigue existiendo, solo pierde su contenido
peliculas.clear()
print(f"Vacia: {peliculas}")
```

</details>

> **Recuerda:** Metodos de listas mas utiles
>
> | Metodo | Descripcion |
> |--------|-------------|
> | `.append(x)` | Agrega `x` al final |
> | `.insert(i, x)` | Inserta `x` en posicion `i` |
> | `.remove(x)` | Elimina la primera aparicion de `x` |
> | `.pop()` | Elimina y devuelve el ultimo (o `.pop(i)` por indice) |
> | `.sort()` | Ordena la lista (modifica el original) |
> | `.reverse()` | Invierte el orden |
> | `.count(x)` | Cuenta apariciones de `x` |
> | `.index(x)` | Posicion de la primera aparicion de `x` |
> | `.clear()` | Vacia la lista |
> | `.copy()` | Crea una copia de la lista |

Indexing y slicing son fundamentales para acceder a elementos de listas en Python.

```python
# --- Ejercicio 4: Indexing y Slicing ---

frutas = ["manzana", "pera", "naranja", "kiwi", "melon"]

# CONCEPTO: Los indices en Python empiezan en 0, no en 1
# Indices positivos:   0         1        2        3       4
# Indices negativos:  -5        -4       -3       -2      -1
# Los negativos cuentan desde el final, lo que evita calcular len()-1

# Acceso por indice positivo: el primer elemento siempre es [0]
print(f"Primero: {frutas[0]}")

# CONCEPTO: Indice -1 siempre apunta al ULTIMO elemento
# Esto es muy util porque no necesitas saber el tamanio de la lista
print(f"Ultimo: {frutas[-1]}")

# --- Slicing: lista[inicio:fin] ---
# CONCEPTO: El indice "fin" NO se incluye en el resultado
# [1:4] devuelve posiciones 1, 2, 3 (tres elementos, no cuatro)
# Esto es por disenio: fin - inicio = cantidad de elementos
print(f"Del 2do al 4to: {frutas[1:4]}")

# --- Slicing con paso negativo ---
# CONCEPTO: [::-1] invierte la lista/string completamente
# Los dos puntos vacios significan "desde el inicio hasta el final"
# El -1 como paso significa "ir hacia atras de uno en uno"
print(f"Al reves: {frutas[::-1]}")

# CONCEPTO: len() devuelve la cantidad de elementos de cualquier coleccion
print(f"Total elementos: {len(frutas)}")

# CONCEPTO: El operador "in" comprueba si un valor EXISTE en la coleccion
# Devuelve True/False. Funciona con listas, tuplas, strings, diccionarios (claves)
print(f"kiwi esta? {'kiwi' in frutas}")
```

</details>

> **Recuerda:** Slicing: `lista[inicio:fin:paso]`
>
> | Expresion | Resultado | Explicacion |
> |-----------|-----------|-------------|
> | `frutas[:]` | Toda la lista (copia) | Sin limites |
> | `frutas[2:]` | Desde posicion 2 al final | Inicio en 2 |
> | `frutas[:3]` | Desde el inicio hasta posicion 2 | Fin NO incluido |
> | `frutas[::2]` | Cada 2 elementos | Paso de 2 |
> | `frutas[::-1]` | Toda la lista al reves | Paso negativo |

> **Ojo:** El indice final **nunca** se incluye. `[1:4]` devuelve posiciones 1, 2 y 3.

Las tuplas son **inmutables**: una vez creadas, no se pueden modificar. Esto las hace utiles para datos fijos.

```python
# --- Ejercicio 5: Tuplas e inmutabilidad ---

# CONCEPTO: Las tuplas se crean con parentesis () y son INMUTABLES
# Inmutable = una vez creada, NO se puede modificar su contenido
# Caso de uso ideal: coordenadas geograficas que nunca cambian
coordenadas = (40.41, -3.70)  # Latitud y longitud de Madrid

# --- Intentar modificar una tupla ---
# CONCEPTO: Si intentas asignar a un indice de tupla, Python lanza TypeError
# coordenadas[0] = 41.0  -> TypeError: 'tuple' object does not support item assignment
# Esto es una PROTECCION: garantiza que los datos no se corrompan accidentalmente

# Acceso por indice funciona igual que en listas (lectura SI, escritura NO)
print(f"Segundo valor: {coordenadas[1]}")

colores = ("rojo", "verde", "azul")  # Tupla de 3 colores

# --- Desempaquetado de tuplas ---
# CONCEPTO: El desempaquetado asigna cada elemento de la tupla a una variable distinta
# La cantidad de variables DEBE coincidir con la cantidad de elementos
# Si no coinciden, Python lanza ValueError
c1, c2, c3 = colores  # c1="rojo", c2="verde", c3="azul"
print(f"c1={c1}, c2={c2}, c3={c3}")
```

</details>

> **Recuerda:** Lista vs Tupla
>
> | Caracteristica | Lista `[]` | Tupla `()` |
> |----------------|-----------|------------|
> | **Mutable** | Si (se puede modificar) | No (inmutable) |
> | **Uso principal** | Datos que cambian | Datos fijos |
> | **Ejemplos** | Lista de compras, notas | Coordenadas, dias de la semana, meses |
> | **Rendimiento** | Ligeramente mas lenta | Ligeramente mas rapida |

La estructura **lista de diccionarios** es muy comun en programacion. Es similar a una tabla o base de datos donde cada diccionario es una "fila" y cada clave es una "columna".

```python
# --- Ejercicio 6: Lista de diccionarios (estructura anidada) ---

# CONCEPTO: Una lista de diccionarios es como una tabla/base de datos
# Cada diccionario es una "fila" y cada clave es una "columna"
# Esta estructura es MUY comun en programacion real (APIs, JSON, bases de datos)
alumnos = [
    {"nombre": "Ana", "edad": 22, "notas": [8.5, 9.0, 7.5]},     # Alumno 0
    {"nombre": "Carlos", "edad": 24, "notas": [6.0, 7.5, 8.0]},   # Alumno 1
    {"nombre": "Maria", "edad": 21, "notas": [9.5, 9.0, 10.0]},   # Alumno 2
]

# --- Acceso anidado: lista[indice]["clave"] ---
# CONCEPTO: Se encadenan los accesos nivel por nivel
# Primero accedemos al diccionario por indice, luego a su clave
# alumnos[1] -> segundo diccionario (Carlos)
# alumnos[1]["nombre"] -> el valor de la clave "nombre" en ese diccionario
print(f"Segundo alumno: {alumnos[1]['nombre']}")

# --- Acceso triple: lista[indice]["clave"][indice] ---
# CONCEPTO: Cuando un valor del diccionario es una lista, necesitamos 3 niveles
# alumnos[2] -> Maria (tercer diccionario)
# alumnos[2]["notas"] -> la lista [9.5, 9.0, 10.0]
# alumnos[2]["notas"][0] -> primer elemento de esa lista (9.5)
print(f"Primera nota del 3ro: {alumnos[2]['notas'][0]}")

# --- Agregar un nuevo diccionario a la lista ---
# CONCEPTO: .append() con un diccionario agrega una nueva "fila" a la "tabla"
alumnos.append({"nombre": "Luis", "edad": 23, "notas": [7.0, 8.5, 6.5]})

# --- Modificar un valor dentro de la estructura anidada ---
# CONCEPTO: Podemos modificar porque tanto la lista como los diccionarios son MUTABLES
alumnos[0]["edad"] = 23  # Cambia la edad de Ana de 22 a 23

# --- Imprimir con calculo de promedio ---
print("\nLista final de alumnos:")
for alumno in alumnos:  # Itera sobre cada diccionario de la lista
    # CONCEPTO: sum() suma todos los elementos de una lista numerica
    # len() cuenta cuantos elementos hay -> la division da el promedio
    promedio = sum(alumno["notas"]) / len(alumno["notas"])
    # CONCEPTO: :.1f formatea el float con 1 decimal (ej: 8.3)
    print(f"  {alumno['nombre']}, {alumno['edad']} anios, "
          f"notas: {alumno['notas']}, promedio: {promedio:.1f}")
```

</details>

> **Recuerda:** Acceso a estructuras anidadas
>
> Cuando combinamos listas y diccionarios, encadenamos los accesos:
>
> ```python
> alumnos[1]['nombre']       # -> "Carlos" (segundo dict, clave 'nombre')
> alumnos[2]['notas'][0]     # -> 9.5 (tercer dict, clave 'notas', primer elemento)
> ```
>
> - Cada `[]` accede a un nivel mas profundo
> - Lista -> se accede por **indice** numerico
> - Diccionario -> se accede por **clave** (string)

```python
# --- Ejercicio 7 (Parte 1): Inventario de tienda - Datos y producto mas caro ---

# CONCEPTO: Lista de diccionarios para modelar datos tabulares
# Cada producto tiene 4 atributos (claves). Todos los productos comparten las mismas claves
# Esto facilita iterar y hacer calculos de forma uniforme
productos = [
    {"nombre": "Laptop",       "precio": 999.99, "stock": 15, "categoria": "Electronica"},
    {"nombre": "Telefono",     "precio": 699.50, "stock": 30, "categoria": "Electronica"},
    {"nombre": "Aspiradora",   "precio": 149.99, "stock": 20, "categoria": "Hogar"},
    {"nombre": "Cafe molido",  "precio": 8.50,   "stock": 100, "categoria": "Alimentacion"},
    {"nombre": "Lampara LED",  "precio": 29.99,  "stock": 45, "categoria": "Hogar"},
]

# --- 1. Encontrar el producto mas caro (patron de busqueda de maximo) ---
# CONCEPTO: Patron clasico de "buscar el maximo":
# 1) Inicializar con el primer elemento como candidato
# 2) Recorrer el resto comparando
# 3) Actualizar el candidato si encontramos uno mejor
producto_caro = productos[0]  # Asumimos que el primero es el mas caro
for producto in productos:
    # Comparamos el precio del producto actual con el del candidato actual
    if producto["precio"] > producto_caro["precio"]:
        producto_caro = producto  # Nuevo candidato: este es mas caro

print(f"Producto mas caro: {producto_caro['nombre']} "
      f"({producto_caro['precio']} EUR)")
```


```python
# --- Ejercicio 7 (Parte 2): Calcular el valor total del inventario ---

# CONCEPTO: Patron de "acumulador": inicializar en 0 y sumar en cada iteracion
# El valor de inventario es: precio * cantidad_en_stock para cada producto
valor_total = 0  # Acumulador inicializado en 0
for producto in productos:
    # Multiplicamos precio unitario por unidades en stock
    valor_producto = producto["precio"] * producto["stock"]
    valor_total += valor_producto  # Equivale a: valor_total = valor_total + valor_producto

# CONCEPTO: :.2f formatea el numero con exactamente 2 decimales
# Importante para cantidades monetarias (siempre mostrar centimos)
print(f"Valor total del inventario: {valor_total:.2f} EUR")
```


```python
# --- Ejercicio 7 (Parte 3): Agrupar productos por categoria ---

# CONCEPTO: Patron de "agrupar por clave" - muy usado en programacion real
# Construimos un diccionario donde:
#   - clave = categoria (string)
#   - valor = lista de nombres de productos de esa categoria
por_categoria = {}  # Diccionario vacio que se ira llenando

for producto in productos:
    cat = producto["categoria"]  # Extraemos la categoria del producto actual

    # CONCEPTO: Comprobamos si la clave ya existe en el diccionario
    # Si NO existe, creamos una lista vacia para esa categoria
    # Si ya existe, no hacemos nada (la lista ya esta creada)
    if cat not in por_categoria:
        por_categoria[cat] = []  # Crear la lista vacia para esta nueva categoria

    # Agregamos el nombre del producto a la lista de su categoria
    por_categoria[cat].append(producto["nombre"])

# --- Imprimir el resultado agrupado ---
print("Productos por categoria:")
# CONCEPTO: .items() nos da pares (clave, valor) para iterar
for categoria, nombres in por_categoria.items():
    print(f"  {categoria}: {nombres}")
```

</details>

> **Recuerda:** Patron de agrupamiento por clave
>
> El patron de **agrupar por clave** es muy comun en programacion real. Los pasos son:
>
> 1. Crear un diccionario vacio
> 2. Para cada elemento, obtener su clave de agrupamiento
> 3. Si esa clave no existe en el diccionario, crear una lista vacia
> 4. Agregar el elemento a la lista de esa clave
>
> ```python
> agrupado = {}
> for item in datos:
>     clave = item["campo"]
>     if clave not in agrupado:
>         agrupado[clave] = []
>     agrupado[clave].append(item)
> ```

Trabajar con matrices como listas de listas nos permite practicar indexing doble y bucles anidados.

```python
# --- Ejercicio 8 (Parte 1): Crear la matriz y acceder a elementos ---

# CONCEPTO: Una matriz se representa como una lista de listas
# Cada lista interna es una FILA de la matriz
# matriz[fila][columna] para acceder a un elemento especifico
matriz = [
    [1, 2, 3],  # Fila 0
    [4, 5, 6],  # Fila 1
    [7, 8, 9]   # Fila 2
]

# Imprimir la matriz fila por fila para visualizarla
print("Matriz original:")
for fila in matriz:  # Cada "fila" es una lista como [1, 2, 3]
    print(f"  {fila}")

# --- Acceso con doble indice ---
# CONCEPTO: matriz[1][2] -> primero accede a la fila 1 ([4,5,6])
# luego accede al elemento en posicion 2 de esa fila (6)
elemento = matriz[1][2]
print(f"\nElemento en fila 1, columna 2: {elemento}")
```


```python
# --- Ejercicio 8 (Parte 2): Suma de cada fila ---

# CONCEPTO: sum() funciona directamente con listas de numeros
# enumerate() nos da el indice de fila para etiquetar cada resultado
print("Suma de cada fila:")
for i, fila in enumerate(matriz):
    # sum(fila) suma todos los elementos de la lista (ej: sum([1,2,3]) = 6)
    print(f"  Fila {i}: {fila} = {sum(fila)}")
```


```python
# --- Ejercicio 8 (Parte 3): Suma de la diagonal principal ---

# CONCEPTO: La diagonal principal de una matriz son los elementos donde fila == columna
# Es decir: [0][0], [1][1], [2][2] -> en este caso: 1, 5, 9
suma_diagonal = 0  # Acumulador para la suma
for i in range(len(matriz)):  # i va de 0 a 2 (3 filas)
    # CONCEPTO: En la diagonal, el indice de fila y columna son IGUALES
    # Por eso usamos matriz[i][i] en vez de matriz[i][j]
    suma_diagonal += matriz[i][i]
print(f"Suma de la diagonal: {suma_diagonal}")
```


```python
# --- Ejercicio 8 (Parte 4): Transponer la matriz ---

# CONCEPTO: Transponer = convertir filas en columnas y viceversa
# El elemento en posicion [i][j] de la original va a [j][i] en la transpuesta
# Original:     Transpuesta:
# 1 2 3         1 4 7
# 4 5 6   ->    2 5 8
# 7 8 9         3 6 9

filas = len(matriz)       # 3 filas en la matriz original
columnas = len(matriz[0]) # 3 columnas (tamanio de la primera fila)

# --- Construir la transpuesta con bucles anidados ---
transpuesta = []  # Lista vacia que contendra las filas de la transpuesta

# CONCEPTO: El bucle EXTERIOR recorre las COLUMNAS de la original (j)
# porque cada columna original se convierte en una fila de la transpuesta
for j in range(columnas):
    nueva_fila = []  # Cada columna original sera una nueva fila

    # El bucle INTERIOR recorre las FILAS de la original (i)
    # recogiendo el elemento [i][j] de cada fila -> forma la nueva fila
    for i in range(filas):
        nueva_fila.append(matriz[i][j])  # Tomamos el elemento de la columna j, fila i

    transpuesta.append(nueva_fila)  # Agregamos la fila completa a la transpuesta

# Imprimir el resultado
print("Matriz transpuesta:")
for fila in transpuesta:
    print(f"  {fila}")
```

</details>

> **Recuerda:** Como funciona la transposicion
>
> - Recorremos las **columnas** de la original (`j = 0, 1, 2`)
> - Para cada columna, recorremos las **filas** (`i = 0, 1, 2`)
> - Tomamos `matriz[i][j]` y lo ponemos en la nueva fila
> - Asi la columna 0 de la original `(1, 4, 7)` se convierte en fila 0 de la transpuesta

> **Tip:** En Python se puede hacer con list comprehension:
> ```python
> transpuesta = [[matriz[i][j] for i in range(filas)] for j in range(columnas)]
> ```
> Esto es mas compacto pero hace exactamente lo mismo.

## Errores Comunes (evita estos!)

| Error | Ejemplo incorrecto | Correccion |
|:---|:---|:---|
| Confundir `[]` (lista) con `()` (tupla) con `{}` (dict) | `mi_tupla = [1, 2, 3]` -- esto es una lista, no tupla | `mi_tupla = (1, 2, 3)` -- parentesis para tuplas, corchetes para listas |
| Modificar una lista mientras la recorres | `for x in lista: lista.remove(x)` -- salta elementos | Crear una copia: `for x in lista[:]: lista.remove(x)` o usar list comprehension |
| Acceder a clave inexistente en diccionario | `agenda["Pedro"]` -> KeyError si no existe | `agenda.get("Pedro")` devuelve `None` sin error, o `agenda.get("Pedro", "default")` |
| Off-by-one con indices negativos | `lista[-0]` esperando el ultimo -> devuelve el primero | `-0` es igual a `0` en Python. Usa `lista[-1]` para el ultimo elemento |
| Confundir `.append()` con `.extend()` | `lista.append([4, 5])` -> agrega `[4, 5]` como UN elemento | `lista.extend([4, 5])` -> agrega `4` y `5` como elementos separados |
