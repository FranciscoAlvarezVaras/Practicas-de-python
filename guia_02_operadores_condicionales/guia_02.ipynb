> **Recuerda:** Operadores Aritmeticos
>
> | Operador | Descripcion | Ejemplo | Resultado |
> |----------|-------------|---------|-----------|

> | `//` | Division entera (redondea hacia -infinito) | `10 // 3` | `3` |
> | `%` | Modulo (resto de la division) | `10 % 3` | `1` |
> | `**` | Potencia | `10 ** 3` | `1000` |

> **Ojo:** La division `/` **siempre** devuelve un `float`, incluso si el resultado es entero (`10 / 2` da `5.0`). La division entera `//` redondea hacia -infinito: `7 // 2 = 3`, `-7 // 2 = -4`.

```python
# --- Solucion Ejercicio 3: Operadores logicos (and, or, not) ---

a = 10

print(f"a>5 and b<5: {a > 5 and b < 5}")

print(f"a<5 or b<5: {a < 5 or b < 5}")

print(f"a>5 and not a==10: {a > 5 and not a == 10}")
```

</details>

> **Recuerda:** Tablas de Verdad
>
> **Operador `and`** (ambas deben ser True):
>
> | A | B | A and B |
> |---|---|--------|
> | True | True | **True** |
> | True | False | False |
> | False | True | False |
> | False | False | False |
>
> **Operador `or`** (al menos una debe ser True):
>
> | A | B | A or B |
> |---|---|-------|
> | True | True | **True** |
> | True | False | **True** |
> | False | True | **True** |
> | False | False | False |
>
> **Operador `not`** (invierte el valor):
>
> | A | not A |
> |---|------|
> | True | False |
> | False | True |

```python
# --- Solucion Ejercicio 4: Comparaciones con tipos mixtos ---

A = 4       # int
B = "Text"  # str
C = 4.1     # float

# Python simplemente devuelve False si los tipos son incompatibles (int vs str)
print(f"A == B: {A == B}")  # int vs str -> False (tipos diferentes, no comparables)

print(f"A > C: {A > C}")    # 4 > 4.1 -> False
print(f"C <= A: {C <= A}")  # 4.1 <= 4 -> False

# != entre str y float -> True (tipos diferentes, nunca son iguales)
# OJO: Si intentaras B > C (str vs float), Python lanzaria TypeError
print(f"B != C: {B != C}")  # str vs float -> True
```

</details>

> **Ojo:** No se puede comparar con `>` o `<` un string con un numero. `B > C` lanzaria un `TypeError`.

`in` y `not in` sirven para buscar si un elemento existe dentro de una coleccion.

```python
# --- Solucion Ejercicio 5: Operadores de pertenencia (in, not in) ---

numeros = [1, 2, 3, 4, 5]  # Lista de enteros

# CONCEPTO: 'in' busca si un elemento EXISTE dentro de una coleccion
# Funciona con listas, tuplas, strings, diccionarios (busca en claves) y sets
print(f"3 in numeros: {3 in numeros}")          # True: 3 esta en la lista

print(f"6 not in numeros: {6 not in numeros}")  # True: 6 no esta en la lista

# Tambien funciona con substrings de varios caracteres: "py" in "python" -> True
print(f"'y' in texto: {'y' in texto}")           # True: 'y' esta en "python"
print(f"'z' not in texto: {'z' not in texto}")    # True: 'z' no esta en "python"
```

</details>

> **Tip:** `in` funciona con: listas, tuplas, strings, diccionarios y sets. Es **muy** util para buscar si algo existe dentro de una coleccion.

Los operadores de asignacion son atajos. En vez de escribir `x = x + 5`, escribes `x += 5`.

```python
# --- Solucion Ejercicio 6: Operadores de asignacion ---

# x += 5 es exactamente igual que x = x + 5, pero mas conciso

# Reseteamos x a 2 para que la potencia de un ejemplo mas claro
x = 2
x **= 3                       # x = x ** 3 = 2^3 = 8
print(f"x **= 3 -> {x}")

print(f"x //= 2 -> {x}")
```

## Seccion 2: Condicionales (`if`, `elif`, `else`)

Los condicionales permiten que tu programa tome decisiones. La estructura basica es:

```python
if condicion1:
    # codigo si condicion1 es True
elif condicion2:
    # codigo si condicion2 es True (y condicion1 era False)
elif condicion3:
    # codigo si condicion3 es True
else:
    # codigo si NINGUNA condicion fue True
```

> **Recuerda:** Python evalua de arriba a abajo y ejecuta el **primer** bloque que sea `True`.

<details>
<summary><b>Ver solucion - Ejercicio 7</b></summary>

El orden de las condiciones importa. Como Python evalua de arriba a abajo, no necesitamos poner el limite inferior en cada `elif` (ya se descarto en la condicion anterior).

```python
# --- Solucion Ejercicio 7: Condicional simple (if/elif/else) ---

nota = 75  # Valor de prueba para la clasificacion

# CONCEPTO: if/elif/else evalua las condiciones de ARRIBA a ABAJO
# y ejecuta SOLO el primer bloque cuya condicion sea True
# Por eso no necesitamos poner el limite inferior: si nota >= 90 fue False,
# sabemos que nota < 90 cuando llegamos al primer elif
if nota >= 90:
    print("Excelente")       # Se ejecuta si nota es 90, 91, 92, ..., 100
elif nota >= 70:
    print("Bueno")           # Se ejecuta si nota es 70-89 (ya descartamos >= 90)
elif nota >= 50:
    print("Suficiente")      # Se ejecuta si nota es 50-69
else:
    print("Insuficiente")    # Se ejecuta si NINGUNA condicion anterior fue True (nota < 50)
```

Para acceder a los datos de un diccionario usamos `p["nombre"]` y `p["edad"]`. Metodos utiles de strings: `.startswith()`, `.endswith()`, `len()`.

```python
# --- Solucion Ejercicio 8: Condicionales con diccionarios ---

# Lista de diccionarios: cada elemento es un dict con claves "nombre" y "edad"
gente = [
    {"nombre": "Jamiro", "edad": 45},
    {"nombre": "Juan",   "edad": 35},
    {"nombre": "Paco",   "edad": 34},
    {"nombre": "Pepe",   "edad": 14},
    {"nombre": "Pilar",  "edad": 24},
    {"nombre": "Laura",  "edad": 24},
    {"nombre": "Jenny",  "edad": 10},
]

# --- 1. Pepe - mayor o menor de edad ---
# CONCEPTO: gente[3] accede al cuarto elemento de la lista (indice 3)
# p["edad"] accede al valor de la clave "edad" dentro del diccionario
p = gente[3]  # p = {"nombre": "Pepe", "edad": 14}
if p["edad"] >= 18:
    print(f"{p['nombre']}: mayor de edad")
else:
    print(f"{p['nombre']}: menor de edad")  # Pepe tiene 14 -> menor

# --- 2. Juan - empieza por J? ---
# CONCEPTO: .startswith() devuelve True si el string empieza por el texto dado
# Es mas legible y seguro que comparar p["nombre"][0] == "J"
p = gente[1]
if p["nombre"].startswith("J"):
    print(f"{p['nombre']}: empieza por J")  # "Juan" empieza por J -> True
else:
    print(f"{p['nombre']}: no empieza por J")

# --- 3. Paco - tiene 4 letras? ---
# CONCEPTO: len() devuelve la longitud del string (numero de caracteres)
p = gente[2]
if len(p["nombre"]) == 4:
    print(f"{p['nombre']}: tiene 4 letras")  # "Paco" tiene 4 letras -> True
else:
    print(f"{p['nombre']}: no tiene 4 letras")

# --- 4. Pilar - edad par o impar? ---
# CONCEPTO: El operador modulo (%) da el resto de la division
# Si el resto de dividir entre 2 es 0, el numero es par
p = gente[4]
if p["edad"] % 2 == 0:
    print(f"{p['nombre']}: edad par ({p['edad']})")    # 24 % 2 = 0 -> par
else:
    print(f"{p['nombre']}: edad impar ({p['edad']})")

# --- 5. Jamiro - termina en "o"? ---
# CONCEPTO: .endswith() devuelve True si el string termina con el texto dado
# Similar a .startswith() pero comprueba el final
p = gente[0]
if p["nombre"].endswith("o"):
    print(f"{p['nombre']}: termina en 'o'")  # "Jamiro" termina en "o" -> True
else:
    print(f"{p['nombre']}: no termina en 'o'")
```

</details>

> **Recuerda:** `p["nombre"]` accede al valor de la clave `"nombre"`, `p["edad"]` accede al valor de la clave `"edad"`.

Combinamos condicionales con operadores logicos. Nota que en Python se pueden encadenar comparaciones: `20 <= x <= 35`.

```python
# --- Solucion Ejercicio 9: Condicionales con operadores logicos ---

gente = [
    {"nombre": "Jamiro", "edad": 45},
    {"nombre": "Juan",   "edad": 35},
    {"nombre": "Paco",   "edad": 34},
    {"nombre": "Pepe",   "edad": 14},
    {"nombre": "Pilar",  "edad": 24},
    {"nombre": "Laura",  "edad": 24},
    {"nombre": "Jenny",  "edad": 10},
]

# --- 1. Juan: empieza por J Y edad < 40 ---
# CONCEPTO: 'and' requiere que AMBAS condiciones sean True
# .startswith("J") -> True, edad < 40 -> True (35) => True and True = True
p = gente[1]
if p["nombre"].startswith("J") and p["edad"] < 40:
    print(f"{p['nombre']}: J y <40")
else:
    print(f"{p['nombre']}: no cumple")

# --- 2. Laura: contiene "a" (case-insensitive) ---
# CONCEPTO: .lower() convierte todo a minusculas ANTES de buscar
# Asi encontramos tanto "a" como "A" sin importar el caso original
p = gente[5]
if "a" in p["nombre"].lower():
    print(f"{p['nombre']}: contiene 'a'")  # "laura" contiene "a" -> True
else:
    print(f"{p['nombre']}: no contiene 'a'")

# --- 3. Pilar: edad entre 20 y 35 ---
# CONCEPTO: Python permite ENCADENAR comparaciones: 20 <= x <= 35
# Esto es equivalente a (x >= 20 and x <= 35) pero mas legible y pythonico
p = gente[4]
if 20 <= p["edad"] <= 35:
    print(f"{p['nombre']}: rango 20-35")  # 24 esta entre 20 y 35 -> True
else:
    print(f"{p['nombre']}: fuera de rango")

# --- 4. Paco: NO empieza por J Y edad > 30 ---
# CONCEPTO: 'not' invierte la condicion. Primero evalua startswith("J") -> False,
# luego 'not False' = True. Combinado con 'and edad > 30' -> True
p = gente[2]
if not p["nombre"].startswith("J") and p["edad"] > 30:
    print(f"{p['nombre']}: no J y >30")  # No empieza por J (True) y 34 > 30 (True)
else:
    print(f"{p['nombre']}: no cumple")

# --- 5. Comparar Juan vs Paco: quien es mayor ---
# Usamos variables 'a' y 'b' (no confundir con los operandos numericos del ej. 1)
a = gente[1]  # Juan, 35
b = gente[2]  # Paco, 34
if a["edad"] > b["edad"]:
    print(f"Mayor: {a['nombre']} ({a['edad']})")  # 35 > 34 -> Juan es mayor
elif b["edad"] > a["edad"]:
    print(f"Mayor: {b['nombre']} ({b['edad']})")
else:
    # CONCEPTO: 'else' cubre todos los casos restantes (aqui, empate exacto)
    print("Empate de edad")
```

</details>

> **Tip:** `.lower()` convierte un string a minusculas, lo que permite buscar sin importar mayusculas/minusculas. En Python puedes encadenar comparaciones: `20 <= x <= 35` es equivalente a `x >= 20 and x <= 35`.

> **Conceptos clave:** `if`/`elif`/`else` | `operaciones aritmeticas` | `f-string formato`

El **Indice de Masa Corporal** (IMC/BMI) se calcula como:

```
IMC = peso_kg / (altura_m ** 2)
```

1. Crea las variables: `peso_kg = 75`, `altura_m = 1.75`
2. Calcula el IMC y guardalo en una variable
3. Imprime el IMC con 1 decimal
4. Clasifica el resultado con `if`/`elif`/`else`:
   - IMC < 18.5: "Bajo peso"
   - 18.5 <= IMC < 25: "Peso normal"
   - 25 <= IMC < 30: "Sobrepeso"
   - IMC >= 30: "Obesidad"
5. Imprime un mensaje completo, ejemplo: `"IMC: 24.5 -> Peso normal"`

### Tu turno

````python
# TU CODIGO AQUI
````

<details>
<summary><b>Ver solucion - Ejercicio 11</b></summary>

Este es un patron muy comun: clasificar un valor numerico en rangos con `if`/`elif`/`else`. No necesitamos el limite inferior en cada `elif` porque si llegamos ahi, la condicion anterior ya fue `False`.

```python
# --- Solucion Ejercicio 11: Calculadora de IMC ---

peso_kg = 75     # Peso en kilogramos
altura_m = 1.75  # Altura en metros

# CONCEPTO: La formula del IMC divide el peso entre la altura al cuadrado
# Los parentesis son NECESARIOS para que ** se aplique solo a altura_m
# Sin parentesis: peso_kg / altura_m ** 2 funcionaria igual por precedencia,
# pero con parentesis queda mas claro y legible
imc = peso_kg / (altura_m ** 2)  # 75 / (1.75^2) = 75 / 3.0625 = 24.49

# --- Clasificar el IMC usando if/elif/else ---
# CONCEPTO: Patron de clasificacion por rangos - el orden importa
# Como evaluamos de arriba a abajo, no necesitamos el limite inferior
# en cada elif (ya quedo descartado por la condicion anterior)
if imc < 18.5:
    categoria = "Bajo peso"        # IMC < 18.5
elif imc < 25:
    categoria = "Peso normal"      # 18.5 <= IMC < 25 (ya sabemos que imc >= 18.5)
elif imc < 30:
    categoria = "Sobrepeso"        # 25 <= IMC < 30
else:
    categoria = "Obesidad"         # IMC >= 30

print(f"Peso: {peso_kg} kg, Altura: {altura_m} m")
# CONCEPTO: :.1f formatea el float mostrando solo 1 decimal
# Es mas legible que mostrar todos los decimales (24.489795918...)
print(f"IMC: {imc:.1f} -> {categoria}")  # IMC: 24.5 -> Peso normal
```

</details>

> **Tip:** `{imc:.1f}` formatea el float con 1 decimal. Con peso=75, altura=1.75: `IMC = 75 / (1.75^2) = 75 / 3.0625 = 24.5` -> "Peso normal".
>
> **Ojo:** El IMC es una medida simplificada. No considera masa muscular, distribucion de grasa ni otros factores importantes.

## Ejercicio 12 de 12: Clasificador de triangulos (MIT - pensamiento geometrico) | Dificil

> **Conceptos clave:** `if`/`elif`/`else` | `sorted()` | `abs()` | `tolerancia floats`

Dados 3 lados de un posible triangulo: `a = 5`, `b = 5`, `c = 5`

**Paso 1:** Verifica si forman un triangulo valido. La suma de cualquier par de lados debe ser **mayor** que el tercer lado:
```
(a + b > c) AND (a + c > b) AND (b + c > a)
```

**Paso 2:** Si NO es valido, imprime: `"No es un triangulo valido"`

**Paso 3:** Si ES valido, clasifica el tipo:
- **Equilatero**: los 3 lados son iguales
- **Isosceles**: exactamente 2 lados son iguales
- **Escaleno**: los 3 lados son diferentes

**Paso 4:** Comprueba si es un **triangulo rectangulo** (Teorema de Pitagoras):
- Ordena los lados de menor a mayor con `sorted([a, b, c])`
- Comprueba si: `menor**2 + medio**2 == mayor**2`
- Usa tolerancia para decimales: `abs(menor**2 + medio**2 - mayor**2) < 0.0001`

**Prueba con estos valores:**

| a | b | c | Resultado esperado |
|---|---|---|--------------------|
| 5 | 5 | 5 | Equilatero |
| 3 | 4 | 5 | Escaleno + Rectangulo |
| 5 | 5 | 3 | Isosceles |
| 1 | 2 | 10 | No valido |

### Tu turno

````python
# TU CODIGO AQUI
````

<details>
<summary><b>Ver solucion - Ejercicio 12</b></summary>

Usamos un bucle `for` para probar varios casos. Conceptos clave:
- `sorted([a, b, c])` devuelve una lista ordenada de menor a mayor
- `continue` salta el resto del bucle y pasa a la siguiente iteracion
- `abs()` devuelve el valor absoluto (convierte negativos a positivos)
- La tolerancia `0.0001` es necesaria porque los floats no son exactos (ej: `0.1 + 0.2 = 0.30000000000000004`)
- `a == b == c`: Python permite encadenar comparaciones de igualdad

```python
# --- Solucion Ejercicio 12: Clasificador de triangulos ---

# Lista de tuplas: cada tupla tiene los 3 lados de un posible triangulo
# Usamos varios casos para probar todos los tipos posibles
casos = [
    (5, 5, 5),    # Caso 1: Equilatero (3 lados iguales)
    (3, 4, 5),    # Caso 2: Escaleno + Rectangulo (pitagorico: 3^2+4^2=5^2)
    (5, 5, 3),    # Caso 3: Isosceles (2 lados iguales)
    (1, 2, 10),   # Caso 4: No valido (1+2=3, no es > 10)
]

# CONCEPTO: 'for a, b, c in casos' desempaqueta cada tupla en 3 variables
# Es equivalente a hacer a, b, c = (5, 5, 5) en cada iteracion
for a, b, c in casos:
    print(f"\nLados: a={a}, b={b}, c={c}")

    # --- Paso 1: Validar si forma un triangulo ---
    # CONCEPTO: Desigualdad triangular - la suma de 2 lados debe ser MAYOR que el tercero
    # Hay que comprobar las 3 combinaciones posibles
    es_valido = (a + b > c) and (a + c > b) and (b + c > a)

    if not es_valido:
        print("  -> No es un triangulo valido")
        # CONCEPTO: 'continue' salta el resto del bucle y pasa a la siguiente iteracion
        # Sin continue, tendriamos que meter todo lo de abajo en un else (menos legible)
        continue

    # --- Paso 2: Clasificar el tipo de triangulo ---
    # CONCEPTO: Python permite encadenar comparaciones de igualdad: a == b == c
    # Es equivalente a (a == b) and (b == c), pero mas legible
    if a == b == c:
        tipo = "Equilatero (3 lados iguales)"
    elif a == b or b == c or a == c:
        # Basta con que UN par de lados sea igual para ser isosceles
        tipo = "Isosceles (2 lados iguales)"
    else:
        # Si ningun par es igual, los 3 son diferentes -> escaleno
        tipo = "Escaleno (3 lados diferentes)"

    print(f"  -> {tipo}")

    # --- Paso 3: Comprobar si es rectangulo (Teorema de Pitagoras) ---
    # CONCEPTO: sorted() devuelve una NUEVA lista ordenada de menor a mayor
    # Necesitamos ordenar para saber cual es la hipotenusa (el lado mayor)
    lados = sorted([a, b, c])
    menor, medio, mayor = lados[0], lados[1], lados[2]  # Desempaquetamos los 3 valores

    # CONCEPTO: Tolerancia para comparar floats
    # Los floats NO son exactos (ej: 0.1 + 0.2 = 0.30000000000000004)
    # Por eso usamos abs(diferencia) < 0.0001 en vez de == directo
    # abs() devuelve el valor absoluto (convierte negativos a positivos)
    if abs(menor**2 + medio**2 - mayor**2) < 0.0001:
        print(f"  -> Es rectangulo! ({menor}^2 + {medio}^2 = "
              f"{menor**2} + {medio**2} = {menor**2 + medio**2} "
              f"= {mayor}^2 = {mayor**2})")
    else:
        print("  -> No es rectangulo")
```

</details>

## Errores Comunes (evita estos!)

| Error | Ejemplo incorrecto | Correccion |
|:---|:---|:---|
| Confundir `=` (asignar) con `==` (comparar) | `if x = 5:` -> SyntaxError | `if x == 5:` -- doble igual para comparar |
| Olvidar los dos puntos `:` despues de `if`/`elif`/`else` | `if x > 5` -> SyntaxError | `if x > 5:` -- siempre poner `:` al final |
| Indentacion incorrecta | Codigo sin sangria dentro del `if` | Usar 4 espacios de indentacion dentro de cada bloque |
| Confundir `and`/`or` en la logica | `if x > 0 or x < 10:` (siempre True) | `if x > 0 and x < 10:` -- usar `and` para "ambas condiciones" |
| Confundir `//` (division entera) con `/` (division float) | `7 / 2` esperando `3` | `7 / 2` da `3.5` (float). Usa `7 // 2` para obtener `3` (entero) |
